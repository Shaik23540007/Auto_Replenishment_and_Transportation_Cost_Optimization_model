```python
import pandas as pd
import pulp as lp
import math

# ==========================================================
# SKU LIST
# ==========================================================

skus = [
    "LG-0090-10AMP",
    "LG-0387-10AMP",
    "LG-0101-10AMP",
    "LG-0013-10MM-CORD",
    "LG-0101-12.5AMP",
    "LG-0090-15AMP",
    "LG-0387-5AMP",
    "LG-0101-5AMP",
    "LG-0060-50W-6K",
    "LG-0080-16W-6K",
    "LG0172-6+3R-6KPGB",
    "LG-0067-12W-RD-3K",
    "LG-0067-12W-RD-4K",
    "LG-0067-12W-RD-6K",
    "LG-0067-12W-SQ-3K",
    "LG-0067-12W-SQ-4K",
    "LG-0067-12W-SQ-6K",
    "LG-0100-15W-RD",
    "LG-0036-10W-6K-3HR",
    "LG-0036-10W-6K-4HR",
    "LG-0025-10W-6K",
    "LG-0457-100-6K"
]
depots = [
    "MYSRD", "BAL", "PHUR", "BUDH", "MUM", "NAG",
    "MILL_RD", "SAN", "CHN", "MUD", "GOL_EXT", "OT",
    "KOL_CITY", "KOL_EXT", "AMD", "HYD", "JPR", "RAP",
    "RAP_EXT", "RAN", "BPL", "GUT", "CUK", "LUK",
    "PAT", "SRN", "DEL", "JAL", "SURAT", "GAZ"
]
warehouses = ["MW1", "MW2"]

# ==========================================================
# SKU MASTER
# ==========================================================
sku_master = {
    "LG-0090-10AMP": {
        "units_per_carton": 100,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Stranger"
    },
    "LG-0387-10AMP": {
        "units_per_carton": 100,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Runner"
    },
    "LG-0101-10AMP": {
        "units_per_carton": 100,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Repeater"
    },
    "LG-0013-10MM-CORD": {
        "units_per_carton": 1,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Stranger"
    },
    "LG-0101-12.5AMP": {
        "units_per_carton": 50,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Stranger"
    },
    "LG-0090-15AMP": {
        "units_per_carton": 50,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Stranger"
    },
    "LG-0387-5AMP": {
        "units_per_carton": 100,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Runner"
    },
    "LG-0101-5AMP": {
        "units_per_carton": 100,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Repeater"
    },
    "LG-0060-50W-6K": {
        "units_per_carton": 20,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Repeater"
    },
    "LG-0080-16W-6K": {
        "units_per_carton": 30,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Repeater"
    },
    "LG0172-6+3R-6KPGB": {
        "units_per_carton": 40,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Repeater"
    },
    "LG-0067-12W-RD-3K": {
        "units_per_carton": 40,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Stranger"
    },
    "LG-0067-12W-RD-4K": {
        "units_per_carton": 40,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Stranger"
    },
    "LG-0067-12W-RD-6K": {
        "units_per_carton": 40,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Repeater"
    },
    "LG-0067-12W-SQ-3K": {
        "units_per_carton": 40,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Stranger"
    },
    "LG-0067-12W-SQ-4K": {
        "units_per_carton": 40,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Stranger"
    },
    "LG-0067-12W-SQ-6K": {
        "units_per_carton": 40,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Stranger"
    },
    "LG-0100-15W-RD": {
        "units_per_carton": 40,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Repeater"
    },
    "LG-0036-10W-6K-3HR": {
        "units_per_carton": 50,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Runner"
    },
    "LG-0036-10W-6K-4HR": {
        "units_per_carton": 50,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Runner"
    },
    "LG-0025-10W-6K": {
        "units_per_carton": 100,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Runner"
    },
    "LG-0457-100-6K": {
        "units_per_carton": 10,
        "actual_wt_carton": 8,
        "length_cm": 40,
        "breadth_cm": 30,
        "height_cm": 25,
        "unit_value": 150,
        "rrs": "Stranger"
    },
}

# ==========================================================
# CBM CALCULATION
# ==========================================================

cbm_per_carton = {}
cbm_per_unit   = {}

for sku in skus:
    sm  = sku_master[sku]
    cpc = (
        sm["length_cm"]
        * sm["breadth_cm"]
        * sm["height_cm"]
    ) / 1_000_000
    cbm_per_carton[sku] = cpc
    cbm_per_unit[sku]   = cpc / sm["units_per_carton"]

# ==========================================================
# BUSINESS PRIORITY CONFIGURATION
# ==========================================================

rrs_penalty = {
    "Runner":   15000,
    "Repeater":  7000,
    "Stranger":  2000
}

depot_weight = {
    "MYSRD":    2.5,
    "BAL":      1.2,
    "PHUR":     1.0,
    "BUDH":     1.0,
    "MUM":      1.3,
    "NAG":      1.0,
    "MILL_RD":  1.0,   # was "MILL RD"
    "SAN":      1.0,
    "CHN":      1.1,
    "MUD":      1.0,
    "GOL_EXT":  2.0,   # was "GOL EXT"
    "OT":       1.0,
    "KOL_CITY": 1.0,   # was "KOL city"
    "KOL_EXT":  1.0,   # was "KOL ext"
    "AMD":      1.0,
    "HYD":      1.5,
    "JPR":      1.0,
    "RAP":      1.0,
    "RAP_EXT":  1.0,
    "RAN":      1.0,
    "BPL":      1.0,
    "GUT":      1.0,
    "CUK":      1.0,
    "LUK":      1.0,
    "PAT":      1.0,
    "SRN":      1.0,
    "DEL":      1.4,
    "JAL":      1.0,
    "SURAT":    1.0,
    "GAZ":      1.0,
}

# ==========================================================
# CURRENT STOCK
# ==========================================================

current_stock = {
    "LG-0090-10AMP": {
        "NEL": 800, "LUH": 600, "MYSRD": 905, "BAL": 86, "PHUR": 273, "BUDH": 114,
        "MUM": 0, "NAG": 136, "MILL_RD": 30, "SAN": 130, "CHN": 127, "MUD": 63,
        "GOL_EXT": 99, "OT": 0, "KOL_CITY": 16, "KOL_EXT": 200, "AMD": 154,
        "HYD": 11, "JPR": 107, "RAP": 3, "RAP_EXT": 0, "RAN": 181, "BPL": 103,
        "GUT": 317, "CUK": 194, "LUK": 153, "PAT": 33, "SRN": 190, "DEL": 687,
        "JAL": 0, "SURAT": 315, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0387-10AMP": {
        "NEL": 18000, "LUH": 21700, "MYSRD": 2125, "BAL": 313, "PHUR": 248, "BUDH": 127,
        "MUM": 281, "NAG": 183, "MILL_RD": 147, "SAN": 1418, "CHN": 1226, "MUD": 955,
        "GOL_EXT": 449, "OT": 131, "KOL_CITY": 307, "KOL_EXT": 500, "AMD": 326,
        "HYD": 165, "JPR": 677, "RAP": 415, "RAP_EXT": 132, "RAN": 391, "BPL": 147,
        "GUT": 1238, "CUK": 541, "LUK": 638, "PAT": 313, "SRN": 500, "DEL": 559,
        "JAL": 200, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0101-10AMP": {
        "NEL": 0, "LUH": 300, "MYSRD": 220, "BAL": 146, "PHUR": 155, "BUDH": 158,
        "MUM": 248, "NAG": 251, "MILL_RD": 72, "SAN": 28, "CHN": 279, "MUD": 134,
        "GOL_EXT": 69, "OT": 130, "KOL_CITY": 183, "KOL_EXT": 100, "AMD": 195,
        "HYD": 184, "JPR": 192, "RAP": 292, "RAP_EXT": 77, "RAN": 453, "BPL": 131,
        "GUT": 338, "CUK": 759, "LUK": 138, "PAT": 293, "SRN": 222, "DEL": 687,
        "JAL": 0, "SURAT": 755, "GAZ": 95, "BHAGIRATH": 0
    },
    "LG-0013-10MM-CORD": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0101-12.5AMP": {
        "NEL": 850, "LUH": 650, "MYSRD": 31, "BAL": 90, "PHUR": 82, "BUDH": 6,
        "MUM": 133, "NAG": 88, "MILL_RD": 16, "SAN": 182, "CHN": 240, "MUD": 131,
        "GOL_EXT": 67, "OT": 62, "KOL_CITY": 85, "KOL_EXT": 50, "AMD": 100,
        "HYD": 113, "JPR": 78, "RAP": 102, "RAP_EXT": 0, "RAN": 76, "BPL": 55,
        "GUT": 88, "CUK": 152, "LUK": 25, "PAT": 74, "SRN": 99, "DEL": 0,
        "JAL": 50, "SURAT": 45, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0090-15AMP": {
        "NEL": 139, "LUH": 200, "MYSRD": 332, "BAL": 1, "PHUR": 40, "BUDH": 19,
        "MUM": 0, "NAG": 30, "MILL_RD": 21, "SAN": 0, "CHN": 0, "MUD": 40,
        "GOL_EXT": 13, "OT": 1, "KOL_CITY": 0, "KOL_EXT": 50, "AMD": 0,
        "HYD": 18, "JPR": 40, "RAP": 40, "RAP_EXT": 0, "RAN": 19, "BPL": 0,
        "GUT": 75, "CUK": 48, "LUK": 40, "PAT": 46, "SRN": 12, "DEL": 235,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0387-5AMP": {
        "NEL": 27200, "LUH": 15900, "MYSRD": 2144, "BAL": 199, "PHUR": 341, "BUDH": 69,
        "MUM": 294, "NAG": 294, "MILL_RD": 171, "SAN": 1603, "CHN": 1340, "MUD": 853,
        "GOL_EXT": 548, "OT": 115, "KOL_CITY": 153, "KOL_EXT": 400, "AMD": 531,
        "HYD": 94, "JPR": 656, "RAP": 390, "RAP_EXT": 340, "RAN": 832, "BPL": 354,
        "GUT": 753, "CUK": 561, "LUK": 2295, "PAT": 225, "SRN": 418, "DEL": 243,
        "JAL": 200, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0101-5AMP": {
        "NEL": 40, "LUH": 300, "MYSRD": 382, "BAL": 127, "PHUR": 259, "BUDH": 66,
        "MUM": 365, "NAG": 373, "MILL_RD": 44, "SAN": 209, "CHN": 218, "MUD": 228,
        "GOL_EXT": 407, "OT": 120, "KOL_CITY": 148, "KOL_EXT": 200, "AMD": 209,
        "HYD": 196, "JPR": 155, "RAP": 200, "RAP_EXT": 193, "RAN": 480, "BPL": 447,
        "GUT": 90, "CUK": 726, "LUK": 200, "PAT": 295, "SRN": 310, "DEL": 899,
        "JAL": 0, "SURAT": 774, "GAZ": 100, "BHAGIRATH": 0
    },
    "LG-0060-50W-6K": {
        "NEL": 1830, "LUH": 680, "MYSRD": 2258, "BAL": 132, "PHUR": 236, "BUDH": 84,
        "MUM": 16, "NAG": 100, "MILL_RD": 57, "SAN": 501, "CHN": 108, "MUD": 395,
        "GOL_EXT": 222, "OT": 69, "KOL_CITY": 95, "KOL_EXT": 360, "AMD": 117,
        "HYD": 103, "JPR": 192, "RAP": 183, "RAP_EXT": 63, "RAN": 395, "BPL": 303,
        "GUT": 286, "CUK": 359, "LUK": 57, "PAT": 197, "SRN": 280, "DEL": 148,
        "JAL": 100, "SURAT": 43, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0080-16W-6K": {
        "NEL": 9600, "LUH": 11610, "MYSRD": 965, "BAL": 258, "PHUR": 1533, "BUDH": 41,
        "MUM": 240, "NAG": 121, "MILL_RD": 100, "SAN": 531, "CHN": 584, "MUD": 391,
        "GOL_EXT": 1050, "OT": 196, "KOL_CITY": 425, "KOL_EXT": 2370, "AMD": 3962,
        "HYD": 435, "JPR": 279, "RAP": 610, "RAP_EXT": 10, "RAN": 210, "BPL": 766,
        "GUT": 465, "CUK": 1159, "LUK": 680, "PAT": 50, "SRN": 331, "DEL": 870,
        "JAL": 300, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG0172-6+3R-6KPGB": {
        "NEL": 1000, "LUH": 5840, "MYSRD": 322, "BAL": 0, "PHUR": 160, "BUDH": 67,
        "MUM": 0, "NAG": 63, "MILL_RD": 0, "SAN": 0, "CHN": 4, "MUD": 8,
        "GOL_EXT": 20, "OT": 1, "KOL_CITY": 80, "KOL_EXT": 200, "AMD": 406,
        "HYD": 0, "JPR": 461, "RAP": 81, "RAP_EXT": 0, "RAN": 374, "BPL": 54,
        "GUT": 158, "CUK": 436, "LUK": 108, "PAT": 360, "SRN": 70, "DEL": 61,
        "JAL": 0, "SURAT": 350, "GAZ": 200, "BHAGIRATH": 0
    },
    "LG-0067-12W-RD-3K": {
        "NEL": 3280, "LUH": 480, "MYSRD": 354, "BAL": 186, "PHUR": 120, "BUDH": 31,
        "MUM": 110, "NAG": 69, "MILL_RD": 38, "SAN": 171, "CHN": 147, "MUD": 70,
        "GOL_EXT": 80, "OT": 25, "KOL_CITY": 85, "KOL_EXT": 120, "AMD": 32,
        "HYD": 40, "JPR": 79, "RAP": 75, "RAP_EXT": 101, "RAN": 0, "BPL": 105,
        "GUT": 75, "CUK": 144, "LUK": 80, "PAT": 18, "SRN": 0, "DEL": 40,
        "JAL": 280, "SURAT": 30, "GAZ": 40, "BHAGIRATH": 80
    },
    "LG-0067-12W-RD-4K": {
        "NEL": 1040, "LUH": 400, "MYSRD": 136, "BAL": 31, "PHUR": 40, "BUDH": 0,
        "MUM": 100, "NAG": 19, "MILL_RD": 28, "SAN": 80, "CHN": 160, "MUD": 24,
        "GOL_EXT": 12, "OT": 27, "KOL_CITY": 54, "KOL_EXT": 80, "AMD": 26,
        "HYD": 35, "JPR": 106, "RAP": 182, "RAP_EXT": 90, "RAN": 0, "BPL": 111,
        "GUT": 46, "CUK": 145, "LUK": 294, "PAT": 80, "SRN": 0, "DEL": 40,
        "JAL": 235, "SURAT": 40, "GAZ": 40, "BHAGIRATH": 0
    },
    "LG-0067-12W-RD-6K": {
        "NEL": 6040, "LUH": 4880, "MYSRD": 1712, "BAL": 144, "PHUR": 232, "BUDH": 108,
        "MUM": 140, "NAG": 67, "MILL_RD": 47, "SAN": 212, "CHN": 178, "MUD": 424,
        "GOL_EXT": 1039, "OT": 178, "KOL_CITY": 85, "KOL_EXT": 280, "AMD": 81,
        "HYD": 253, "JPR": 131, "RAP": 384, "RAP_EXT": 222, "RAN": 138, "BPL": 214,
        "GUT": 951, "CUK": 951, "LUK": 1624, "PAT": 229, "SRN": 118, "DEL": 258,
        "JAL": 208, "SURAT": 118, "GAZ": 200, "BHAGIRATH": 67
    },
    "LG-0067-12W-SQ-3K": {
        "NEL": 160, "LUH": 0, "MYSRD": 225, "BAL": 142, "PHUR": 20, "BUDH": 48,
        "MUM": 87, "NAG": 74, "MILL_RD": 44, "SAN": 40, "CHN": 55, "MUD": 55,
        "GOL_EXT": 20, "OT": 43, "KOL_CITY": 40, "KOL_EXT": 40, "AMD": 50,
        "HYD": 83, "JPR": 119, "RAP": 63, "RAP_EXT": 15, "RAN": 0, "BPL": 78,
        "GUT": 25, "CUK": 95, "LUK": 23, "PAT": 20, "SRN": 0, "DEL": 0,
        "JAL": 440, "SURAT": 40, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0067-12W-SQ-4K": {
        "NEL": 145, "LUH": 40, "MYSRD": 56, "BAL": 91, "PHUR": 41, "BUDH": 33,
        "MUM": 57, "NAG": 39, "MILL_RD": 2, "SAN": 15, "CHN": 59, "MUD": 41,
        "GOL_EXT": 16, "OT": 55, "KOL_CITY": 47, "KOL_EXT": 40, "AMD": 63,
        "HYD": 13, "JPR": 51, "RAP": 7, "RAP_EXT": 109, "RAN": 30, "BPL": 114,
        "GUT": 18, "CUK": 160, "LUK": 190, "PAT": 40, "SRN": 0, "DEL": 0,
        "JAL": 240, "SURAT": 40, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0067-12W-SQ-6K": {
        "NEL": 4480, "LUH": 2560, "MYSRD": 551, "BAL": 143, "PHUR": 146, "BUDH": 122,
        "MUM": 160, "NAG": 126, "MILL_RD": 33, "SAN": 145, "CHN": 91, "MUD": 204,
        "GOL_EXT": 182, "OT": 77, "KOL_CITY": 81, "KOL_EXT": 120, "AMD": 171,
        "HYD": 74, "JPR": 97, "RAP": 241, "RAP_EXT": 80, "RAN": 40, "BPL": 62,
        "GUT": 43, "CUK": 335, "LUK": 462, "PAT": 80, "SRN": 26, "DEL": 40,
        "JAL": 209, "SURAT": 200, "GAZ": 40, "BHAGIRATH": 200
    },
    "LG-0100-15W-RD": {
        "NEL": 2080, "LUH": 800, "MYSRD": 418, "BAL": 72, "PHUR": 258, "BUDH": 49,
        "MUM": 198, "NAG": 138, "MILL_RD": 18, "SAN": 141, "CHN": 630, "MUD": 135,
        "GOL_EXT": 306, "OT": 170, "KOL_CITY": 94, "KOL_EXT": 40, "AMD": 62,
        "HYD": 125, "JPR": 69, "RAP": 154, "RAP_EXT": 241, "RAN": 50, "BPL": 109,
        "GUT": 0, "CUK": 234, "LUK": 160, "PAT": 99, "SRN": 0, "DEL": 79,
        "JAL": 160, "SURAT": 200, "GAZ": 0, "BHAGIRATH": 111
    },
    "LG-0036-10W-6K-3HR": {
        "NEL": 0, "LUH": 0, "MYSRD": 15254, "BAL": 3, "PHUR": 0, "BUDH": 82,
        "MUM": 26, "NAG": 593, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 406, "KOL_EXT": 300, "AMD": 552,
        "HYD": 65, "JPR": 0, "RAP": 10, "RAP_EXT": 8, "RAN": 200, "BPL": 329,
        "GUT": 542, "CUK": 353, "LUK": 23, "PAT": 0, "SRN": 289, "DEL": 275,
        "JAL": 0, "SURAT": 2, "GAZ": 0, "BHAGIRATH": 946
    },
    "LG-0036-10W-6K-4HR": {
        "NEL": 50, "LUH": 0, "MYSRD": 99, "BAL": 67, "PHUR": 5950, "BUDH": 1792,
        "MUM": 602, "NAG": 5488, "MILL_RD": 28, "SAN": 15, "CHN": 611, "MUD": 6,
        "GOL_EXT": 0, "OT": 14, "KOL_CITY": 3157, "KOL_EXT": 16291, "AMD": 0,
        "HYD": 69, "JPR": 124, "RAP": 1049, "RAP_EXT": 858, "RAN": 251, "BPL": 4996,
        "GUT": 43, "CUK": 5760, "LUK": 0, "PAT": 5706, "SRN": 1408, "DEL": 724,
        "JAL": 550, "SURAT": 578, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0025-10W-6K": {
        "NEL": 676300, "LUH": 699898, "MYSRD": 86053, "BAL": 21501, "PHUR": 138404, "BUDH": 1720,
        "MUM": 16355, "NAG": 23032, "MILL_RD": 2523, "SAN": 149678, "CHN": 15224, "MUD": 36255,
        "GOL_EXT": 221174, "OT": 3923, "KOL_CITY": 9215, "KOL_EXT": 52100, "AMD": 179981,
        "HYD": 75097, "JPR": 25153, "RAP": 115218, "RAP_EXT": 8240, "RAN": 64600, "BPL": 114067,
        "GUT": 122149, "CUK": 36075, "LUK": 69244, "PAT": 134690, "SRN": 8270, "DEL": 44785,
        "JAL": 200, "SURAT": 63909, "GAZ": 34590, "BHAGIRATH": 6600
    },
    "LG-0457-100-6K": {
        "NEL": 13490, "LUH": 3570, "MYSRD": 105, "BAL": 28, "PHUR": 256, "BUDH": 11,
        "MUM": 422, "NAG": 44, "MILL_RD": 36, "SAN": 298, "CHN": 101, "MUD": 108,
        "GOL_EXT": 217, "OT": 55, "KOL_CITY": 8, "KOL_EXT": 0, "AMD": 0,
        "HYD": 190, "JPR": 204, "RAP": 149, "RAP_EXT": 168, "RAN": 101, "BPL": 101,
        "GUT": 0, "CUK": 388, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 399,
        "JAL": 374, "SURAT": 100, "GAZ": 399, "BHAGIRATH": 200
    },
}

# ==========================================================
# FREEZE STOCK
# ==========================================================

freeze_stock = {
    "LG-0013-10MM-CORD": {
        "NEL": 0, "LUH": 0, "MYSRD": 1000, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0090-10AMP": {
        "NEL": 209, "LUH": 109, "MYSRD": 300, "BAL": 110, "PHUR": 300, "BUDH": 0,
        "MUM": 0, "NAG": 200, "MILL_RD": 0, "SAN": 200, "CHN": 200, "MUD": 110,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 200, "AMD": 200,
        "HYD": 0, "JPR": 110, "RAP": 0, "RAP_EXT": 0, "RAN": 200, "BPL": 0,
        "GUT": 200, "CUK": 200, "LUK": 0, "PAT": 0, "SRN": 200, "DEL": 200,
        "JAL": 110, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0090-15AMP": {
        "NEL": 32, "LUH": 14, "MYSRD": 15, "BAL": 10, "PHUR": 20, "BUDH": 0,
        "MUM": 0, "NAG": 20, "MILL_RD": 0, "SAN": 15, "CHN": 0, "MUD": 15,
        "GOL_EXT": 15, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 20, "AMD": 0,
        "HYD": 0, "JPR": 20, "RAP": 0, "RAP_EXT": 0, "RAN": 20, "BPL": 0,
        "GUT": 20, "CUK": 0, "LUK": 0, "PAT": 20, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0060-50W-6K": {
        "NEL": 3000, "LUH": 2000, "MYSRD": 2780, "BAL": 300, "PHUR": 300, "BUDH": 0,
        "MUM": 0, "NAG": 100, "MILL_RD": 0, "SAN": 500, "CHN": 140, "MUD": 400,
        "GOL_EXT": 240, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 300, "AMD": 120,
        "HYD": 140, "JPR": 200, "RAP": 200, "RAP_EXT": 0, "RAN": 400, "BPL": 300,
        "GUT": 300, "CUK": 600, "LUK": 160, "PAT": 200, "SRN": 0, "DEL": 140,
        "JAL": 100, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0080-16W-6K": {
        "NEL": 6500, "LUH": 10000, "MYSRD": 1020, "BAL": 0, "PHUR": 2500, "BUDH": 0,
        "MUM": 180, "NAG": 120, "MILL_RD": 0, "SAN": 500, "CHN": 210, "MUD": 300,
        "GOL_EXT": 900, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 2100, "AMD": 4000,
        "HYD": 900, "JPR": 300, "RAP": 600, "RAP_EXT": 0, "RAN": 300, "BPL": 800,
        "GUT": 800, "CUK": 900, "LUK": 570, "PAT": 210, "SRN": 120, "DEL": 120,
        "JAL": 1000, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG0172-6+3R-6KPGB": {
        "NEL": 700, "LUH": 3300, "MYSRD": 360, "BAL": 0, "PHUR": 240, "BUDH": 0,
        "MUM": 0, "NAG": 120, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 90, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 800, "AMD": 400,
        "HYD": 80, "JPR": 800, "RAP": 80, "RAP_EXT": 0, "RAN": 400, "BPL": 80,
        "GUT": 300, "CUK": 1000, "LUK": 50, "PAT": 80, "SRN": 80, "DEL": 80,
        "JAL": 80, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0067-12W-RD-3K": {
        "NEL": 1200, "LUH": 450, "MYSRD": 360, "BAL": 120, "PHUR": 50, "BUDH": 0,
        "MUM": 80, "NAG": 60, "MILL_RD": 0, "SAN": 160, "CHN": 80, "MUD": 80,
        "GOL_EXT": 120, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 120, "AMD": 45,
        "HYD": 80, "JPR": 80, "RAP": 80, "RAP_EXT": 0, "RAN": 0, "BPL": 120,
        "GUT": 112, "CUK": 80, "LUK": 50, "PAT": 50, "SRN": 0, "DEL": 0,
        "JAL": 50, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0067-12W-RD-4K": {
        "NEL": 500, "LUH": 550, "MYSRD": 160, "BAL": 50, "PHUR": 50, "BUDH": 0,
        "MUM": 80, "NAG": 80, "MILL_RD": 0, "SAN": 80, "CHN": 60, "MUD": 60,
        "GOL_EXT": 90, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 80, "AMD": 45,
        "HYD": 80, "JPR": 80, "RAP": 80, "RAP_EXT": 0, "RAN": 0, "BPL": 120,
        "GUT": 80, "CUK": 80, "LUK": 50, "PAT": 50, "SRN": 0, "DEL": 0,
        "JAL": 50, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0067-12W-RD-6K": {
        "NEL": 6200, "LUH": 5000, "MYSRD": 2000, "BAL": 200, "PHUR": 240, "BUDH": 0,
        "MUM": 80, "NAG": 400, "MILL_RD": 0, "SAN": 280, "CHN": 120, "MUD": 520,
        "GOL_EXT": 800, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 600, "AMD": 80,
        "HYD": 720, "JPR": 120, "RAP": 600, "RAP_EXT": 0, "RAN": 120, "BPL": 200,
        "GUT": 1500, "CUK": 800, "LUK": 50, "PAT": 160, "SRN": 0, "DEL": 80,
        "JAL": 80, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0067-12W-SQ-3K": {
        "NEL": 300, "LUH": 150, "MYSRD": 120, "BAL": 50, "PHUR": 50, "BUDH": 0,
        "MUM": 0, "NAG": 60, "MILL_RD": 0, "SAN": 120, "CHN": 50, "MUD": 60,
        "GOL_EXT": 60, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 80, "AMD": 45,
        "HYD": 0, "JPR": 50, "RAP": 50, "RAP_EXT": 0, "RAN": 0, "BPL": 80,
        "GUT": 50, "CUK": 50, "LUK": 50, "PAT": 50, "SRN": 0, "DEL": 0,
        "JAL": 50, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0067-12W-SQ-4K": {
        "NEL": 200, "LUH": 150, "MYSRD": 20, "BAL": 10, "PHUR": 20, "BUDH": 0,
        "MUM": 80, "NAG": 20, "MILL_RD": 0, "SAN": 30, "CHN": 15, "MUD": 15,
        "GOL_EXT": 15, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 20,
        "HYD": 0, "JPR": 20, "RAP": 20, "RAP_EXT": 0, "RAN": 20, "BPL": 20,
        "GUT": 20, "CUK": 40, "LUK": 0, "PAT": 20, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0067-12W-SQ-6K": {
        "NEL": 2200, "LUH": 1500, "MYSRD": 640, "BAL": 120, "PHUR": 240, "BUDH": 0,
        "MUM": 80, "NAG": 120, "MILL_RD": 0, "SAN": 200, "CHN": 80, "MUD": 240,
        "GOL_EXT": 240, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 120, "AMD": 120,
        "HYD": 80, "JPR": 120, "RAP": 320, "RAP_EXT": 0, "RAN": 50, "BPL": 120,
        "GUT": 120, "CUK": 200, "LUK": 50, "PAT": 80, "SRN": 0, "DEL": 80,
        "JAL": 50, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0100-15W-RD": {
        "NEL": 2600, "LUH": 1800, "MYSRD": 520, "BAL": 80, "PHUR": 500, "BUDH": 0,
        "MUM": 120, "NAG": 400, "MILL_RD": 0, "SAN": 120, "CHN": 240, "MUD": 240,
        "GOL_EXT": 400, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 120, "AMD": 80,
        "HYD": 380, "JPR": 80, "RAP": 480, "RAP_EXT": 0, "RAN": 50, "BPL": 120,
        "GUT": 80, "CUK": 240, "LUK": 50, "PAT": 80, "SRN": 0, "DEL": 80,
        "JAL": 80, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0457-100-6K": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0036-10W-6K-3HR": {
        "NEL": 50000, "LUH": 30000, "MYSRD": 15000, "BAL": 1000, "PHUR": 5000, "BUDH": 0,
        "MUM": 350, "NAG": 300, "MILL_RD": 0, "SAN": 3000, "CHN": 500, "MUD": 2000,
        "GOL_EXT": 7500, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 2500, "AMD": 500,
        "HYD": 5000, "JPR": 1000, "RAP": 4500, "RAP_EXT": 0, "RAN": 500, "BPL": 5000,
        "GUT": 2000, "CUK": 1000, "LUK": 1000, "PAT": 1000, "SRN": 300, "DEL": 350,
        "JAL": 1000, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0036-10W-6K-4HR": {
        "NEL": 100000, "LUH": 75000, "MYSRD": 15000, "BAL": 500, "PHUR": 20000, "BUDH": 0,
        "MUM": 750, "NAG": 3000, "MILL_RD": 0, "SAN": 5000, "CHN": 3000, "MUD": 10000,
        "GOL_EXT": 10000, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 5000, "AMD": 6000,
        "HYD": 5000, "JPR": 2000, "RAP": 3500, "RAP_EXT": 0, "RAN": 3500, "BPL": 5000,
        "GUT": 3000, "CUK": 5000, "LUK": 2000, "PAT": 3000, "SRN": 300, "DEL": 400,
        "JAL": 500, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0025-10W-6K": {
        "NEL": 1000000, "LUH": 1000000, "MYSRD": 100000, "BAL": 6000, "PHUR": 0, "BUDH": 0,
        "MUM": 12200, "NAG": 5000, "MILL_RD": 0, "SAN": 0, "CHN": 20000, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 80000,
        "HYD": 0, "JPR": 30000, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 50000, "CUK": 0, "LUK": 50000, "PAT": 0, "SRN": 6000, "DEL": 18000,
        "JAL": 30000, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0387-10AMP": {
        "NEL": 13000, "LUH": 8500, "MYSRD": 2600, "BAL": 200, "PHUR": 500, "BUDH": 0,
        "MUM": 300, "NAG": 200, "MILL_RD": 0, "SAN": 2000, "CHN": 400, "MUD": 1000,
        "GOL_EXT": 600, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 500, "AMD": 400,
        "HYD": 800, "JPR": 600, "RAP": 500, "RAP_EXT": 0, "RAN": 600, "BPL": 300,
        "GUT": 1500, "CUK": 1000, "LUK": 900, "PAT": 300, "SRN": 300, "DEL": 300,
        "JAL": 800, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0387-5AMP": {
        "NEL": 14000, "LUH": 10000, "MYSRD": 2900, "BAL": 200, "PHUR": 500, "BUDH": 0,
        "MUM": 300, "NAG": 300, "MILL_RD": 0, "SAN": 2500, "CHN": 300, "MUD": 1000,
        "GOL_EXT": 600, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 700, "AMD": 600,
        "HYD": 700, "JPR": 600, "RAP": 400, "RAP_EXT": 0, "RAN": 1000, "BPL": 500,
        "GUT": 1000, "CUK": 1000, "LUK": 2500, "PAT": 400, "SRN": 300, "DEL": 400,
        "JAL": 1200, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0101-10AMP": {
        "NEL": 3500, "LUH": 2000, "MYSRD": 500, "BAL": 200, "PHUR": 500, "BUDH": 0,
        "MUM": 200, "NAG": 200, "MILL_RD": 0, "SAN": 1000, "CHN": 300, "MUD": 300,
        "GOL_EXT": 600, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 400, "AMD": 200,
        "HYD": 600, "JPR": 300, "RAP": 300, "RAP_EXT": 0, "RAN": 500, "BPL": 300,
        "GUT": 400, "CUK": 500, "LUK": 110, "PAT": 300, "SRN": 300, "DEL": 200,
        "JAL": 300, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0101-12.5AMP": {
        "NEL": 800, "LUH": 300, "MYSRD": 150, "BAL": 60, "PHUR": 100, "BUDH": 0,
        "MUM": 100, "NAG": 60, "MILL_RD": 0, "SAN": 200, "CHN": 100, "MUD": 100,
        "GOL_EXT": 150, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 100, "AMD": 100,
        "HYD": 100, "JPR": 100, "RAP": 60, "RAP_EXT": 0, "RAN": 100, "BPL": 60,
        "GUT": 60, "CUK": 100, "LUK": 60, "PAT": 100, "SRN": 60, "DEL": 0,
        "JAL": 60, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0101-5AMP": {
        "NEL": 3500, "LUH": 2600, "MYSRD": 600, "BAL": 200, "PHUR": 500, "BUDH": 0,
        "MUM": 200, "NAG": 300, "MILL_RD": 0, "SAN": 1000, "CHN": 300, "MUD": 320,
        "GOL_EXT": 600, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 400, "AMD": 300,
        "HYD": 500, "JPR": 300, "RAP": 300, "RAP_EXT": 0, "RAN": 500, "BPL": 500,
        "GUT": 200, "CUK": 700, "LUK": 110, "PAT": 500, "SRN": 300, "DEL": 200,
        "JAL": 200, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
}

# ==========================================================
# LAST 4 MONTH SALES HISTORY
# ==========================================================

sales_history = {
    "LG-0090-10AMP": {
        "MYSRD": [18, 42, 72],   "BAL": [8, 19, 33],    "PHUR": [6, 14, 24],
        "BUDH":  [4,  9, 16],    "MUM": [5, 12, 20],    "NAG":  [4, 10, 17],
        "MILL_RD":[3, 7, 12],    "SAN": [5, 11, 19],    "CHN":  [7, 16, 27],
        "MUD":   [3,  8, 13],    "GOL_EXT":[4,9,16],    "OT":   [2,  5,  9],
        "KOL_CITY":[2,5,8],      "KOL_EXT":[3,7,12],    "AMD":  [5, 12, 20],
        "HYD":   [4,  9, 16],    "JPR":  [4, 10, 17],   "RAP":  [2,  4,  7],
        "RAP_EXT":[1, 3,  5],    "RAN":  [3,  8, 13],   "BPL":  [3,  7, 12],
        "GUT":   [5, 12, 20],    "CUK":  [4, 10, 17],   "LUK":  [4,  9, 15],
        "PAT":   [2,  5,  9],    "SRN":  [3,  7, 12],   "DEL":  [7, 16, 27],
        "JAL":   [1,  2,  4],    "SURAT":[4, 10, 17],   "GAZ":  [1,  2,  4],
    },
    "LG-0387-10AMP": {
        "MYSRD": [480,1380,2580], "BAL": [75,215,402],   "PHUR": [59,170,318],
        "BUDH":  [30, 87, 163],   "MUM": [67,193,361],   "NAG":  [44,126,235],
        "MILL_RD":[35,101,189],   "SAN": [339,975,1823], "CHN": [293,842,1574],
        "MUD":  [228,656,1226],   "GOL_EXT":[107,308,576],"OT": [31, 90, 168],
        "KOL_CITY":[73,211,394],  "KOL_EXT":[120,344,644],"AMD":[78,224,419],
        "HYD":  [39, 113, 211],   "JPR": [162,465,870],  "RAP": [99, 285,533],
        "RAP_EXT":[32,91,170],    "RAN": [93, 268,502],  "BPL": [35,101,189],
        "GUT":  [296,851,1591],   "CUK": [129,372,695],  "LUK": [152,438,819],
        "PAT":  [75, 215,402],    "SRN": [120,344,644],  "DEL": [134,385,719],
        "JAL":  [48, 138,258],    "SURAT":[0,  0,  0],   "GAZ": [0,   0,  0],
    },
    "LG-0101-10AMP": {
        "MYSRD": [30, 88, 164],   "BAL": [20, 57, 106],  "PHUR": [21, 61, 114],
        "BUDH":  [22, 62, 116],   "MUM": [34, 97, 181],  "NAG":  [34, 98, 183],
        "MILL_RD":[10,29, 54],    "SAN": [4,  11,  20],  "CHN":  [38,109, 204],
        "MUD":   [18, 53,  99],   "GOL_EXT":[9,27,50],   "OT":   [18, 52,  97],
        "KOL_CITY":[25,72,135],   "KOL_EXT":[14,39,73],  "AMD":  [27, 77, 143],
        "HYD":   [25, 72, 135],   "JPR": [26, 75, 140],  "RAP":  [40,115, 215],
        "RAP_EXT":[11,31, 58],    "RAN": [62,178, 332],  "BPL":  [18, 52,  97],
        "GUT":   [46,133, 249],   "CUK": [104,299,559],  "LUK":  [19, 54, 101],
        "PAT":   [40,115, 215],   "SRN": [30, 87, 163],  "DEL":  [94,271, 506],
        "JAL":   [0,   0,   0],   "SURAT":[103,298,557], "GAZ":  [13, 38,  71],
    },
    "LG-0013-10MM-CORD": {
        "MYSRD": [0, 0, 0],  "BAL": [0, 0, 0],  "PHUR": [0, 0, 0],
        "BUDH":  [0, 0, 0],  "MUM": [0, 0, 0],  "NAG":  [0, 0, 0],
        "MILL_RD":[0,0, 0],  "SAN": [0, 0, 0],  "CHN":  [0, 0, 0],
        "MUD":   [0, 0, 0],  "GOL_EXT":[0,0,0], "OT":   [0, 0, 0],
        "KOL_CITY":[0,0,0],  "KOL_EXT":[0,0,0], "AMD":  [0, 0, 0],
        "HYD":   [0, 0, 0],  "JPR": [0, 0, 0],  "RAP":  [0, 0, 0],
        "RAP_EXT":[0,0, 0],  "RAN": [0, 0, 0],  "BPL":  [0, 0, 0],
        "GUT":   [0, 0, 0],  "CUK": [0, 0, 0],  "LUK":  [0, 0, 0],
        "PAT":   [0, 0, 0],  "SRN": [0, 0, 0],  "DEL":  [0, 0, 0],
        "JAL":   [0, 0, 0],  "SURAT":[0,0, 0],  "GAZ":  [0, 0, 0],
    },
    "LG-0101-12.5AMP": {
        "MYSRD": [4,  12,  22],   "BAL": [12, 34,  64],  "PHUR": [11, 32,  59],
        "BUDH":  [1,   2,   4],   "MUM": [18, 52,  97],  "NAG":  [12, 34,  64],
        "MILL_RD":[2,  6,  11],   "SAN": [25, 71, 133],  "CHN":  [33, 94, 176],
        "MUD":   [18, 51,  95],   "GOL_EXT":[9,26,49],   "OT":   [8,  24,  45],
        "KOL_CITY":[12,33,62],    "KOL_EXT":[7,19,36],   "AMD":  [14, 39,  73],
        "HYD":   [15, 44,  82],   "JPR": [11, 30,  57],  "RAP":  [14, 40,  75],
        "RAP_EXT":[0,  0,   0],   "RAN": [10, 30,  56],  "BPL":  [7,  22,  41],
        "GUT":   [12, 34,  64],   "CUK": [21, 59, 110],  "LUK":  [3,  10,  18],
        "PAT":   [10, 29,  54],   "SRN": [14, 39,  73],  "DEL":  [0,   0,   0],
        "JAL":   [7,  19,  36],   "SURAT":[6, 18,  33],  "GAZ":  [0,   0,   0],
    },
    "LG-0090-15AMP": {
        "MYSRD": [7,  8,  14],    "BAL": [0,  0,   1],   "PHUR": [2,  2,   7],
        "BUDH":  [1,  1,   3],    "MUM": [0,  0,   0],   "NAG":  [1,  2,   5],
        "MILL_RD":[1, 1,   4],    "SAN": [0,  0,   0],   "CHN":  [0,  0,   0],
        "MUD":   [1,  2,   7],    "GOL_EXT":[0,1,2],     "OT":   [0,  0,   0],
        "KOL_CITY":[0, 0,  0],    "KOL_EXT":[1,1,2],     "AMD":  [0,  0,   0],
        "HYD":   [1,  1,   3],    "JPR": [1,  2,   7],   "RAP":  [1,  2,   7],
        "RAP_EXT":[0, 0,   0],    "RAN": [1,  1,   3],   "BPL":  [0,  0,   0],
        "GUT":   [2,  2,  13],    "CUK": [1,  2,   8],   "LUK":  [1,  2,   7],
        "PAT":   [2,  2,   8],    "SRN": [0,  1,   2],   "DEL":  [2,  4,   6],
        "JAL":   [0,  0,   0],    "SURAT":[0, 0,   0],   "GAZ":  [0,  0,   0],
    },
    "LG-0387-5AMP": {
        "MYSRD": [550,1580,2880],  "BAL": [51,146,273],  "PHUR": [87,251,469],
        "BUDH":  [18, 51,  95],    "MUM": [75,216,404],  "NAG":  [75,216,404],
        "MILL_RD":[44,126,235],    "SAN": [410,1179,2205],"CHN": [343,986,1843],
        "MUD":  [218,628,1174],    "GOL_EXT":[140,403,753],"OT": [29, 85, 158],
        "KOL_CITY":[39,113,211],   "KOL_EXT":[102,295,551],"AMD":[136,391,731],
        "HYD":  [24,  69, 128],    "JPR": [168,482,901],  "RAP": [100,287,536],
        "RAP_EXT":[87,251,469],    "RAN": [213,612,1144], "BPL": [91,261,488],
        "GUT":  [193,555,1037],    "CUK": [144,413,772],  "LUK": [587,1690,3159],
        "PAT":  [58, 166,310],     "SRN": [107,307,574],  "DEL": [62, 179,335],
        "JAL":  [51, 147,275],     "SURAT":[0,  0,  0],   "GAZ": [0,   0,  0],
    },
    "LG-0101-5AMP": {
        "MYSRD": [52, 150,281],   "BAL": [17, 50,  94],  "PHUR": [35, 101,189],
        "BUDH":  [9,  26,  48],   "MUM": [50, 143,267],  "NAG":  [51, 146,273],
        "MILL_RD":[6,  17, 32],   "SAN": [28, 82, 153],  "CHN":  [30, 86, 160],
        "MUD":   [31, 90, 168],   "GOL_EXT":[56,160,300],"OT":   [16, 47,  88],
        "KOL_CITY":[20,58,108],   "KOL_EXT":[27,79,148], "AMD":  [28, 82, 153],
        "HYD":   [27, 77, 143],   "JPR": [21, 61, 114],  "RAP":  [27, 78, 146],
        "RAP_EXT":[26,76,142],    "RAN": [65,188, 352],  "BPL":  [61,175, 328],
        "GUT":   [12, 35,  66],   "CUK": [99,285, 533],  "LUK":  [27, 78, 146],
        "PAT":   [40,115, 215],   "SRN": [42,121, 227],  "DEL":  [123,353,660],
        "JAL":   [0,   0,   0],   "SURAT":[106,304,569], "GAZ":  [14,  39,  73],
    },
    "LG-0060-50W-6K": {
        "MYSRD": [484,526,250],   "BAL": [22, 24,  11],  "PHUR": [39, 42,  20],
        "BUDH":  [14, 15,   7],   "MUM": [3,   3,   1],  "NAG":  [16, 18,   8],
        "MILL_RD":[9,  10,  5],   "SAN": [82, 89,  42],  "CHN":  [18, 19,   9],
        "MUD":   [65, 70,  33],   "GOL_EXT":[36,40,19],  "OT":   [11, 12,   6],
        "KOL_CITY":[16,17, 8],    "KOL_EXT":[59,64,30],  "AMD":  [19, 21,  10],
        "HYD":   [17, 18,   9],   "JPR": [31, 34,  16],  "RAP":  [30, 33,  15],
        "RAP_EXT":[10,11,  5],    "RAN": [65, 70,  33],  "BPL":  [50, 54,  25],
        "GUT":   [47, 51,  24],   "CUK": [59, 64,  30],  "LUK":  [9,  10,   5],
        "PAT":   [32, 35,  16],   "SRN": [46, 50,  23],  "DEL":  [24, 26,  12],
        "JAL":   [16, 18,   8],   "SURAT":[7,  8,   4],  "GAZ":  [0,   0,   0],
    },
    "LG-0080-16W-6K": {
        "MYSRD": [318,344,157],   "BAL": [84, 91,  41],  "PHUR": [503,544,248],
        "BUDH":  [13, 15,   7],   "MUM": [79, 85,  39],  "NAG":  [40, 43,  19],
        "MILL_RD":[33,35,  16],   "SAN": [174,188,  86], "CHN":  [191,207,  94],
        "MUD":   [128,139,  63],  "GOL_EXT":[344,373,170],"OT":  [64, 70,  32],
        "KOL_CITY":[139,151,69],  "KOL_EXT":[777,841,384],"AMD": [1298,1405,641],
        "HYD":   [143,154,  70],  "JPR": [91,  99,  45],  "RAP": [200,216,  99],
        "RAP_EXT":[3,  4,   2],   "RAN": [69,  74,  34],  "BPL": [251,272, 124],
        "GUT":   [152,165,  75],  "CUK": [380,411, 188],  "LUK": [223,241, 110],
        "PAT":   [16, 18,   8],   "SRN": [108,117,  54],  "DEL": [285,309, 141],
        "JAL":   [98,107,  49],   "SURAT":[0,  0,   0],   "GAZ": [0,   0,   0],
    },
    "LG0172-6+3R-6KPGB": {
        "MYSRD": [159,183,118],   "BAL": [0,   0,   0],  "PHUR": [79, 91,  59],
        "BUDH":  [33, 38,  24],   "MUM": [0,   0,   0],  "NAG":  [31, 36,  23],
        "MILL_RD":[0,  0,   0],   "SAN": [0,   0,   0],  "CHN":  [2,   2,   1],
        "MUD":   [4,   4,   3],   "GOL_EXT":[10,11,7],   "OT":   [0,   0,   0],
        "KOL_CITY":[40,46,29],    "KOL_EXT":[99,114,74], "AMD":  [201,232, 150],
        "HYD":   [0,   0,   0],   "JPR": [228,263, 170], "RAP":  [40,  46,  30],
        "RAP_EXT":[0,  0,   0],   "RAN": [185,214, 138], "BPL":  [27,  31,  20],
        "GUT":   [78,  90,  58],  "CUK": [216,249, 161], "LUK":  [53,  62,  40],
        "PAT":   [178,206, 133],  "SRN": [35,  40,  26], "DEL":  [30,  35,  22],
        "JAL":   [0,   0,   0],   "SURAT":[173,200,129], "GAZ":  [99, 114,  74],
    },
    "LG-0067-12W-RD-3K": {
        "MYSRD": [63, 73,  51],   "BAL": [37, 43,  30],  "PHUR": [24, 27,  19],
        "BUDH":  [6,   7,   5],   "MUM": [22, 25,  18],  "NAG":  [14, 16,  11],
        "MILL_RD":[8,  9,   6],   "SAN": [34, 40,  28],  "CHN":  [29, 34,  24],
        "MUD":   [14, 16,  11],   "GOL_EXT":[16,18,13],  "OT":   [5,   6,   4],
        "KOL_CITY":[17,20,14],    "KOL_EXT":[24,28,19],  "AMD":  [6,   8,   5],
        "HYD":   [8,   9,   6],   "JPR": [16, 18,  13],  "RAP":  [15, 17,  12],
        "RAP_EXT":[20,23,16],     "RAN": [0,   0,   0],  "BPL":  [21, 24,  17],
        "GUT":   [15, 17,  12],   "CUK": [29, 33,  23],  "LUK":  [16, 18,  13],
        "PAT":   [4,   4,   3],   "SRN": [0,   0,   0],  "DEL":  [8,   9,   6],
        "JAL":   [56, 65,  45],   "SURAT":[6,  7,   5],  "GAZ":  [8,   9,   6],
    },
    "LG-0067-12W-RD-4K": {
        "MYSRD": [27, 23,  18],   "BAL": [6,   5,   4],  "PHUR": [8,   6,   5],
        "BUDH":  [0,   0,   0],   "MUM": [20, 16,  12],  "NAG":  [4,   3,   2],
        "MILL_RD":[6,  4,   3],   "SAN": [16, 13,  10],  "CHN":  [32, 26,  19],
        "MUD":   [5,   4,   3],   "GOL_EXT":[2,2,1],     "OT":   [5,   4,   3],
        "KOL_CITY":[11,8,  6],    "KOL_EXT":[16,13,10],  "AMD":  [5,   4,   3],
        "HYD":   [7,   6,   4],   "JPR": [21, 17,  13],  "RAP":  [36, 29,  22],
        "RAP_EXT":[18,15,11],     "RAN": [0,   0,   0],  "BPL":  [22, 18,  13],
        "GUT":   [9,   8,   6],   "CUK": [29, 23,  17],  "LUK":  [59, 47,  35],
        "PAT":   [16, 13,  10],   "SRN": [0,   0,   0],  "DEL":  [8,   6,   5],
        "JAL":   [47, 38,  28],   "SURAT":[8,  6,   5],  "GAZ":  [8,   6,   5],
    },
    "LG-0067-12W-RD-6K": {
        "MYSRD": [547,646,309],   "BAL": [46, 54,  26],  "PHUR": [74, 87,  42],
        "BUDH":  [34, 41,  19],   "MUM": [45, 53,  25],  "NAG":  [21, 25,  12],
        "MILL_RD":[15,18,  8],    "SAN": [68, 80,  38],  "CHN":  [57, 67,  32],
        "MUD":   [135,160, 76],   "GOL_EXT":[332,391,187],"OT":  [57, 67,  32],
        "KOL_CITY":[27,32,15],    "KOL_EXT":[89,106,50], "AMD":  [26, 30,  14],
        "HYD":   [81, 95,  45],   "JPR": [42, 49,  24],  "RAP":  [123,144,  69],
        "RAP_EXT":[71,83, 40],    "RAN": [44, 52,  25],  "BPL":  [68, 81,  38],
        "GUT":   [304,358,171],   "CUK": [304,358, 171], "LUK":  [519,612, 293],
        "PAT":   [73, 86,  41],   "SRN": [38, 44,  21],  "DEL":  [82, 97,  46],
        "JAL":   [66, 78,  37],   "SURAT":[38,44,  21],  "GAZ":  [64, 75,  36],
    },
    "LG-0067-12W-SQ-3K": {
        "MYSRD": [45, 71,  47],   "BAL": [28, 45,  30],  "PHUR": [4,   6,   4],
        "BUDH":  [10, 15,  10],   "MUM": [17, 28,  18],  "NAG":  [15, 23,  16],
        "MILL_RD":[9, 14,   9],   "SAN": [8,  13,   9],  "CHN":  [11, 17,  12],
        "MUD":   [11, 17,  12],   "GOL_EXT":[4,6,4],     "OT":   [9,  14,   9],
        "KOL_CITY":[8,13,  8],    "KOL_EXT":[8,13, 8],   "AMD":  [10, 16,  10],
        "HYD":   [17, 26,  17],   "JPR": [24, 38,  25],  "RAP":  [13, 20,  13],
        "RAP_EXT":[3,  5,   3],   "RAN": [0,   0,   0],  "BPL":  [16, 25,  16],
        "GUT":   [5,   8,   5],   "CUK": [19, 30,  20],  "LUK":  [5,   7,   5],
        "PAT":   [4,   6,   4],   "SRN": [0,   0,   0],  "DEL":  [0,   0,   0],
        "JAL":   [88,139,  92],   "SURAT":[8, 13,   8],  "GAZ":  [0,   0,   0],
    },
    "LG-0067-12W-SQ-4K": {
        "MYSRD": [11, 5,  20],    "BAL": [18,  8,  31],  "PHUR": [8,   4,  14],
        "BUDH":  [7,   3,  11],   "MUM": [11,  5,  20],  "NAG":  [8,   3,  13],
        "MILL_RD":[0,  0,   1],   "SAN": [3,   1,   5],  "CHN":  [12,  5,  20],
        "MUD":   [8,   4,  14],   "GOL_EXT":[3,1,5],     "OT":   [11,  5,  19],
        "KOL_CITY":[9, 4, 16],    "KOL_EXT":[8,4,14],    "AMD":  [13,  6,  22],
        "HYD":   [3,   1,   4],   "JPR": [10,  5,  18],  "RAP":  [1,   1,   2],
        "RAP_EXT":[22,10,  37],   "RAN": [6,   3,  10],  "BPL":  [23, 10,  39],
        "GUT":   [4,   2,   6],   "CUK": [32, 14,  55],  "LUK":  [38, 17,  65],
        "PAT":   [8,   4,  14],   "SRN": [0,   0,   0],  "DEL":  [0,   0,   0],
        "JAL":   [48, 21,  83],   "SURAT":[8,  4,  14],  "GAZ":  [0,   0,   0],
    },
    "LG-0067-12W-SQ-6K": {
        "MYSRD": [176,137,  75],  "BAL": [46, 35,  19],  "PHUR": [47, 36,  20],
        "BUDH":  [39, 30,  17],   "MUM": [51, 40,  22],  "NAG":  [40, 31,  17],
        "MILL_RD":[11,8,    4],   "SAN": [46, 36,  20],  "CHN":  [29, 23,  12],
        "MUD":   [65, 51,  28],   "GOL_EXT":[58,45,25],  "OT":   [25, 19,  10],
        "KOL_CITY":[26,20,11],    "KOL_EXT":[38,30,16],  "AMD":  [55, 42,  23],
        "HYD":   [24, 18,  10],   "JPR": [31, 24,  13],  "RAP":  [77, 60,  33],
        "RAP_EXT":[26,20,  11],   "RAN": [13, 10,   5],  "BPL":  [20, 15,   8],
        "GUT":   [14, 11,   6],   "CUK": [107,83,  46],  "LUK":  [148,115,  63],
        "PAT":   [26, 20,  11],   "SRN": [8,   6,   3],  "DEL":  [13, 10,   5],
        "JAL":   [67, 52,  28],   "SURAT":[64,50,  27],  "GAZ":  [13, 10,   5],
    },
    "LG-0100-15W-RD": {
        "MYSRD": [134,172,108],   "BAL": [23, 30,  18],  "PHUR": [83, 106,  67],
        "BUDH":  [16, 20,  13],   "MUM": [64, 81,  51],  "NAG":  [44, 57,  36],
        "MILL_RD":[6,  7,   5],   "SAN": [45, 58,  37],  "CHN":  [202,259, 163],
        "MUD":   [43, 56,  35],   "GOL_EXT":[98,126,79], "OT":   [55, 70,  44],
        "KOL_CITY":[30,39,24],    "KOL_EXT":[13,16,10],  "AMD":  [20, 26,  16],
        "HYD":   [40, 51,  32],   "JPR": [22, 28,  18],  "RAP":  [49, 63,  40],
        "RAP_EXT":[77,99,  62],   "RAN": [16, 21,  13],  "BPL":  [35, 45,  28],
        "GUT":   [0,   0,   0],   "CUK": [75, 96,  60],  "LUK":  [51, 66,  41],
        "PAT":   [32, 41,  26],   "SRN": [0,   0,   0],  "DEL":  [25, 33,  21],
        "JAL":   [51, 66,  41],   "SURAT":[64,82,  51],  "GAZ":  [0,   0,   0],
    },
    "LG-0036-10W-6K-3HR": {
        "MYSRD": [0,   0,   0],   "BAL": [0,   1,   0],  "PHUR": [0,   0,   0],
        "BUDH":  [13, 14,   2],   "MUM": [4,   5,   1],  "NAG":  [96, 104,  18],
        "MILL_RD":[0,  0,   0],   "SAN": [0,   0,   0],  "CHN":  [0,   0,   0],
        "MUD":   [0,   0,   0],   "GOL_EXT":[0, 0,  0],  "OT":   [0,   0,   0],
        "KOL_CITY":[66,71, 13],   "KOL_EXT":[49,53,9],   "AMD":  [89,  97,  17],
        "HYD":   [11, 11,   2],   "JPR": [0,   0,   0],  "RAP":  [2,   2,   0],
        "RAP_EXT":[1,  1,   0],   "RAN": [32, 35,   6],  "BPL":  [53,  58,  10],
        "GUT":   [88, 95,  17],   "CUK": [57, 62,  11],  "LUK":  [4,   4,   1],
        "PAT":   [0,   0,   0],   "SRN": [47, 51,   9],  "DEL":  [45,  48,   9],
        "JAL":   [0,   0,   0],   "SURAT":[0,  0,   0],  "GAZ":  [0,   0,   0],
    },
    "LG-0036-10W-6K-4HR": {
        "MYSRD": [3,  14,   8],   "BAL": [2,  10,   5],  "PHUR": [188,909, 515],
        "BUDH":  [57,274, 155],   "MUM": [19, 92,  52],  "NAG":  [174,839,475],
        "MILL_RD":[1,  4,   2],   "SAN": [0,   2,   1],  "CHN":  [19, 93,  53],
        "MUD":   [0,   1,   0],   "GOL_EXT":[0,  0,  0], "OT":   [0,   1,   0],
        "KOL_CITY":[100,483,274], "KOL_EXT":[515,2487,1410],"AMD":[0,   0,   0],
        "HYD":   [2,  11,   6],   "JPR": [4,  19,  11],  "RAP":  [33, 161,  91],
        "RAP_EXT":[27,131,  74],  "RAN": [8,  38,  22],  "BPL":  [158,765, 433],
        "GUT":   [1,   7,   4],   "CUK": [182,881,499],  "LUK":  [0,   0,   0],
        "PAT":   [180,872, 494],  "SRN": [44, 215, 122], "DEL":  [23, 110,  62],
        "JAL":   [17,  84,  47],  "SURAT":[18,89,  51],  "GAZ":  [0,   0,   0],
    },
    "LG-0025-10W-6K": {
        "MYSRD": [2756,6082,1283], "BAL": [690,1523,321],  "PHUR": [4446,9813,2070],
        "BUDH":  [55,  122,  26],  "MUM": [525,1160,245],  "NAG":  [740,1633,344],
        "MILL_RD":[81, 179,  38],  "SAN": [4807,10612,2239],"CHN": [489,1079,228],
        "MUD":  [1164,2570, 542],  "GOL_EXT":[7108,15693,3311],"OT":[126,278,59],
        "KOL_CITY":[296,653,138], "KOL_EXT":[1674,3695,780], "AMD":[5781,12763,2692],
        "HYD":  [2412,5326,1123], "JPR": [808,1784,376],   "RAP": [3701,8170,1724],
        "RAP_EXT":[265,585,123],  "RAN": [2075,4581,967],  "BPL": [3665,8091,1707],
        "GUT":  [3924,8664,1828], "CUK": [1159,2559,540],  "LUK": [2225,4912,1036],
        "PAT":  [4328,9553,2015], "SRN": [266,587,124],    "DEL": [1439,3177,670],
        "JAL":  [6,   14,   3],   "SURAT":[2053,4533,956], "GAZ": [1111,2453,518],
    },
    "LG-0457-100-6K": {
        "MYSRD": [5,   3,   1],   "BAL": [1,   1,   0],  "PHUR": [10,  6,   3],
        "BUDH":  [0,   0,   0],   "MUM": [17,  10,   4], "NAG":  [2,   1,   1],
        "MILL_RD":[1,  1,   0],   "SAN": [12,  7,   3],  "CHN":  [4,   2,   1],
        "MUD":   [4,   3,   1],   "GOL_EXT":[9,5,2],     "OT":   [2,   1,   1],
        "KOL_CITY":[0, 0,   0],   "KOL_EXT":[0,0,  0],   "AMD":  [0,   0,   0],
        "HYD":   [8,   4,   2],   "JPR": [8,   5,   2],  "RAP":  [6,   3,   1],
        "RAP_EXT":[7,  4,   2],   "RAN": [4,   2,   1],  "BPL":  [4,   2,   1],
        "GUT":   [0,   0,   0],   "CUK": [16,  9,   4],  "LUK":  [0,   0,   0],
        "PAT":   [0,   0,   0],   "SRN": [0,   0,   0],  "DEL":  [16,  9,   4],
        "JAL":   [15,  9,   4],   "SURAT":[4,  2,   1],  "GAZ":  [16,  9,   4],
    },
}

# ==========================================================
# SUPPLIER INBOUND
# ==========================================================

supplier_inbound = {
    sku: {d: 0 for d in depots}
    for sku in skus
}

# ==========================================================
# INTRANSIT
# ==========================================================

intransit = {
    "LG-0090-10AMP": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0387-10AMP": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 300, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 100,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 200,
        "GUT": 100, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0101-10AMP": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 100, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0013-10MM-CORD": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0101-12.5AMP": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 50, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 100, "BHAGIRATH": 0
    },
    "LG-0090-15AMP": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 40, "BHAGIRATH": 0
    },
    "LG-0387-5AMP": {
        "NEL": 0, "LUH": 0, "MYSRD": 200, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 300, "MUD": 400,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 200,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 200,
        "GUT": 200, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0101-5AMP": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 100,
        "HYD": 0, "JPR": 100, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 200, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0060-50W-6K": {
        "NEL": 0, "LUH": 0, "MYSRD": 40, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 60, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0080-16W-6K": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 60, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 90,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 180,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG0172-6+3R-6KPGB": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 40, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 40,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 40,
        "GUT": 40, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0067-12W-RD-3K": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 64, "BHAGIRATH": 0
    },
    "LG-0067-12W-RD-4K": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 80, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 65, "BHAGIRATH": 0
    },
    "LG-0067-12W-RD-6K": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 320, "MILL_RD": 0, "SAN": 0, "CHN": 40, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 200, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 280,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0067-12W-SQ-3K": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 40, "BHAGIRATH": 0
    },
    "LG-0067-12W-SQ-4K": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0067-12W-SQ-6K": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 120,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 74, "BHAGIRATH": 0
    },
    "LG-0100-15W-RD": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 120, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 400, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 120,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0036-10W-6K-3HR": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0036-10W-6K-4HR": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 1950, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 800, "MUD": 6500,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 3200,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0025-10W-6K": {
        "NEL": 0, "LUH": 0, "MYSRD": 6500, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 0, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 0,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 0, "BHAGIRATH": 0
    },
    "LG-0457-100-6K": {
        "NEL": 0, "LUH": 0, "MYSRD": 0, "BAL": 0, "PHUR": 0, "BUDH": 0,
        "MUM": 0, "NAG": 0, "MILL_RD": 0, "SAN": 0, "CHN": 10, "MUD": 0,
        "GOL_EXT": 0, "OT": 0, "KOL_CITY": 0, "KOL_EXT": 0, "AMD": 10,
        "HYD": 0, "JPR": 0, "RAP": 0, "RAP_EXT": 0, "RAN": 0, "BPL": 0,
        "GUT": 0, "CUK": 0, "LUK": 0, "PAT": 0, "SRN": 0, "DEL": 0,
        "JAL": 0, "SURAT": 0, "GAZ": 3, "BHAGIRATH": 0
    },
}

# ==========================================================
# WAREHOUSE STOCK
# ==========================================================

warehouse_stock = {
    "MW1": {  # NEL
        "LG-0090-10AMP":      800,
        "LG-0387-10AMP":    18000,
        "LG-0101-10AMP":        0,
        "LG-0013-10MM-CORD":    100000,
        "LG-0101-12.5AMP":    850,
        "LG-0090-15AMP":      139,
        "LG-0387-5AMP":     27200,
        "LG-0101-5AMP":        40,
        "LG-0060-50W-6K":    1830,
        "LG-0080-16W-6K":    9600,
        "LG0172-6+3R-6KPGB": 1000,
        "LG-0067-12W-RD-3K": 3280,
        "LG-0067-12W-RD-4K": 1040,
        "LG-0067-12W-RD-6K": 6040,
        "LG-0067-12W-SQ-3K":  160,
        "LG-0067-12W-SQ-4K":  145,
        "LG-0067-12W-SQ-6K": 4480,
        "LG-0100-15W-RD":    2080,
        "LG-0036-10W-6K-3HR":   0,
        "LG-0036-10W-6K-4HR":  50,
        "LG-0025-10W-6K":  676300,
        "LG-0457-100-6K":   13490,
    },

    "MW2": {  # LUH
        "LG-0090-10AMP":      600,
        "LG-0387-10AMP":    21700,
        "LG-0101-10AMP":      300,
        "LG-0013-10MM-CORD":    10000,
        "LG-0101-12.5AMP":    650,
        "LG-0090-15AMP":      200,
        "LG-0387-5AMP":     15900,
        "LG-0101-5AMP":       300,
        "LG-0060-50W-6K":     680,
        "LG-0080-16W-6K":   11610,
        "LG0172-6+3R-6KPGB": 5840,
        "LG-0067-12W-RD-3K":  480,
        "LG-0067-12W-RD-4K":  400,
        "LG-0067-12W-RD-6K": 4880,
        "LG-0067-12W-SQ-3K":    0,
        "LG-0067-12W-SQ-4K":   40,
        "LG-0067-12W-SQ-6K": 2560,
        "LG-0100-15W-RD":     800,
        "LG-0036-10W-6K-3HR":   0,
        "LG-0036-10W-6K-4HR":   0,
        "LG-0025-10W-6K":  699898,
        "LG-0457-100-6K":    3570,
    },
}

# ==========================================================
# MOQ
# ==========================================================

MOQ_WD_CARTONS = {
    (w, d, sku): 2
    for w in warehouses
    for d in depots
    for sku in skus
}

# ==========================================================
# VEHICLE MASTER
# ==========================================================

vehicles = ["20FT", "22FT", "24FT", "32FT"]

vehicle_capacity_cbm = {
    "20FT": 28,
    "22FT": 32,
    "24FT": 38,
    "32FT": 52
}

vehicle_hire_cost = {
    "20FT": 18000,
    "22FT": 20000,
    "24FT": 23000,
    "32FT": 31000
}

# ==========================================================
# DISTANCE MATRIX
# ==========================================================

dist_data = {
    #         MW1   MW2   NEL   LUH  MYSRD   BAL  PHUR  BUDH   MUM   NAG  MILL_RD   SAN   CHN   MUD GOLEXT    OT  KOLC  KOLX   AMD   HYD   JPR   RAP RAPXT   RAN   BPL   GUT   CUK   LUK   PAT   SRN   DEL   JAL SURAT   GAZ  BHAG
    "MW1":  [   0,  120, 1700, 1800,  850,  1600, 2100, 1900,  1400, 1100,  1200, 2100, 2300, 1900,  1800, 1700, 1500, 1500,  980,  1500, 1100,  900,  950,  1300, 1050,  1400, 1500, 1350, 1500, 1600,  550,  950,  1200, 1050, 1800],
    "MW2":  [ 120,    0, 1750, 1900,  920,  1680, 2200, 2000,  1480, 1150,  1280, 2180, 2380, 1980,  1900, 1800, 1580, 1580, 1050,  1580, 1180,  980, 1020,  1380, 1130,  1480, 1550, 1430, 1580, 1680,  620, 1020,  1260, 1130, 1870],
    "NEL":  [1700, 1750,    0,  120,  1600, 2400, 2900, 2700,  2200, 1900,  2000, 2900, 3100, 2700,  2600, 2500, 2300, 2300, 1800,  2300, 1900, 1700, 1750,  2100, 1850,  2200, 2300, 2150, 2300, 2400, 1750, 1750,  2000, 1850, 2600],
    "LUH":  [1800, 1900,  120,    0,  1700, 2500, 3000, 2800,  2300, 2000,  2100, 3000, 3200, 2800,  2700, 2600, 2400, 2400, 1900,  2400, 2000, 1800, 1850,  2200, 1950,  2300, 2400, 2250, 2400, 2500, 1850, 1850,  2100, 1950, 2700],
    "MYSRD":[850,  920, 1600, 1700,     0,   750, 1350, 1100,   600,  550,   600, 1350, 1550, 1150,  1050,  950,  750,  750,  350,   750,  800,  600,  650,   950,  700,  1050, 1150, 1000, 1150, 1200,  500,  700,   700,  750, 1350],
    "BAL":  [1600, 1680, 2400, 2500,  750,     0,  600,  400,   900, 1200,  1100,  900,  900, 1200,  1100, 1000, 1400, 1400, 1200,  1000, 1500, 1300, 1350,  1650, 1400,  1100,  850,  700,  850,  950, 1700, 1650,  1350, 1400,  900],
    "PHUR": [2100, 2200, 2900, 3000, 1350,  600,     0,  200,  1400, 1700,  1600,  700,  900, 1700,  1600, 1500, 1900, 1900, 1700,   900, 2000, 1800, 1850,  2150, 1900,  1600, 1400, 1200, 1400, 1550, 2300, 2150,  1900, 1900, 1500],
    "BUDH": [1900, 2000, 2700, 2800, 1100,  400,  200,     0,  1200, 1500,  1400,  900,  700, 1500,  1400, 1300, 1700, 1700, 1500,   700, 1800, 1600, 1650,  1950, 1700,  1400, 1200, 1000, 1200, 1350, 2100, 1950,  1650, 1700, 1300],
    "MUM":  [1400, 1480, 2200, 2300,  600,  900, 1400, 1200,      0,  750,   650, 1350, 1350,  750,   750,  850, 1350, 1350, 1100,  1600,  1400, 1200, 1250,  1550, 1300,   750,  600,  850,  600,  700, 1700, 1400,  1100, 1300, 1350],
    "NAG":  [1100, 1150, 1900, 2000,  550, 1200, 1700, 1500,   750,    0,   100, 1600, 1800, 1100,  1000,  900,  700,  700,  800,  1700,  1100,  900,  950,  1250, 1000,   900, 1100, 1100,  900,  950, 1400, 1100,   900,  950, 1600],
    "MILL_RD":[1200,1280,2000,2100,  600, 1100, 1600, 1400,  650,  100,     0, 1550, 1700, 1100,  1000,  900,  800,  800,  850,  1650,  1200,  950, 1000,  1300, 1050,   900, 1100, 1100,  900,  950, 1500, 1150,  1000, 1000, 1550],
    "SAN":  [2100, 2180, 2900, 3000, 1350,  900,  700,  900,  1350, 1600,  1550,    0,  400, 1600,  1500, 1400, 1800, 1800, 1600,   500, 1900, 1700, 1750,  2050, 1800,  1500, 1300, 1100, 1300, 1450, 2200, 2050,  1750, 1800, 1500],
    "CHN":  [2300, 2380, 3100, 3200, 1550,  900,  900,  700,  1350, 1800,  1700,  400,    0, 1800,  1700, 1600, 2000, 2000, 1800,   350, 2100, 1900, 1950,  2250, 2000,  1700, 1500, 1300, 1500, 1650, 2400, 2250,  1950, 2000, 1700],
    "MUD":  [1900, 1980, 2700, 2800, 1150, 1200, 1700, 1500,   750, 1100,  1100, 1600, 1800,    0,   200,  300, 1100, 1100, 1300,  1700,  1600, 1400, 1450,  1750, 1500,   900, 1000, 1200, 1000, 1050, 2200, 1750,  1300, 1400, 1800],
    "GOL_EXT":[1800,1900,2600,2700,1050,1100,1600,1400,  750, 1000,  1000, 1500, 1700,  200,     0,  150, 1000, 1000, 1200,  1600,  1500, 1300, 1350,  1650, 1400,   800,  900, 1100,  900,  950, 2100, 1650,  1200, 1300, 1700],
    "OT":   [1700, 1800, 2500, 2600,  950, 1000, 1500, 1300,  850,  900,   900, 1400, 1600,  300,  150,     0,  900,  900, 1100,  1500,  1400, 1200, 1250,  1550, 1300,   700,  800, 1000,  800,  850, 2000, 1550,  1100, 1200, 1600],
    "KOL_CITY":[1500,1580,2300,2400, 750,1400,1900,1700, 1350,  700,   800, 1800, 2000, 1100, 1000,  900,     0,   20, 1600,  1900,  1800, 1600, 1650,  1950, 1700,  1200, 1300, 1500, 1300, 1350, 1800, 1800,  1500, 1600, 2000],
    "KOL_EXT":[1500,1580,2300,2400,  750,1400,1900,1700, 1350,  700,   800, 1800, 2000, 1100, 1000,  900,   20,    0, 1600,  1900,  1800, 1600, 1650,  1950, 1700,  1200, 1300, 1500, 1300, 1350, 1800, 1800,  1500, 1600, 2000],
    "AMD":  [ 980, 1050, 1800, 1900,  350, 1200, 1700, 1500,  1100, 800,   850, 1600, 1800, 1300,  1200, 1100, 1600, 1600,     0,  1700,  1200, 1000, 1050,  1350, 1100,  1100, 1300, 1200, 1300, 1350,  900, 1050,   600, 1000, 1600],
    "HYD":  [1500, 1580, 2300, 2400,  750, 1000,  900,  700,  1600, 1700,  1650,  500,  350, 1700,  1600, 1500, 1900, 1900, 1700,     0, 1900, 1700, 1750,  2050, 1800,  1500, 1300, 1100, 1300, 1450, 2100, 1950,  1600, 1700, 1300],
    "JPR":  [1100, 1180, 1900, 2000,  800, 1500, 2000, 1800,  1400, 1100,  1200, 1900, 2100, 1600,  1500, 1400, 1800, 1800, 1200,  1900,     0,  350,  400,   700,  450,  1200, 1400, 1250, 1400, 1450,  600,  600,   600,  550, 1900],
    "RAP":  [ 900,  980, 1700, 1800,  600, 1300, 1800, 1600,  1200,  900,   950, 1700, 1900, 1400,  1300, 1200, 1600, 1600, 1000,  1700,  350,    0,  100,   400,  200,  1000, 1200, 1100, 1200, 1250,  400,  400,   400,  350, 1700],
    "RAP_EXT":[950,1020,1750,1850,  650,1350,1850,1650, 1250,  950,  1000, 1750, 1950, 1450, 1350, 1250, 1650, 1650, 1050, 1750,  400,  100,     0,   350,  250,  1050, 1250, 1100, 1250, 1300,  450,  450,   450,  400, 1750],
    "RAN":  [1300, 1380, 2100, 2200,  950, 1650, 2150, 1950,  1550, 1250,  1300, 2050, 2250, 1750,  1650, 1550, 1950, 1950, 1350,  2050,  700,  400,  350,     0,  300,  1300, 1500, 1350, 1500, 1550,  700,  700,   700,  650, 2050],
    "BPL":  [1050, 1130, 1850, 1950,  700, 1400, 1900, 1700,  1300, 1000,  1050, 1800, 2000, 1500,  1400, 1300, 1700, 1700, 1100,  1800,  450,  200,  250,   300,    0,  1050, 1250, 1100, 1250, 1300,  500,  500,   550,  500, 1800],
    "GUT":  [1400, 1480, 2200, 2300, 1050, 1100, 1600, 1400,   750,  900,   900, 1500, 1700,  900,   800,  700, 1200, 1200, 1100,  1500,  1200, 1000, 1050,  1300, 1050,     0,  500,  600,  500,  600, 1700, 1250,  1050, 1050, 1500],
    "CUK":  [1500, 1550, 2300, 2400, 1150,  850, 1400, 1200,   600, 1100,  1100, 1300, 1500, 1000,   900,  800, 1300, 1300, 1300,  1300,  1400, 1200, 1250,  1500, 1250,  500,     0,  350,  300,  400, 1800, 1400,  1200, 1250, 1300],
    "LUK":  [1350, 1430, 2150, 2250, 1000,  700, 1200, 1000,   850,  1100, 1100, 1100, 1300, 1200,  1100, 1000, 1500, 1500, 1200,  1100,  1250, 1100, 1100,  1350, 1100,  600,  350,     0,  350,  450, 1700, 1300,  1150, 1100, 1100],
    "PAT":  [1500, 1580, 2300, 2400, 1150,  850, 1400, 1200,   600, 900,    900, 1300, 1500, 1000,   900,  800, 1300, 1300, 1300,  1300,  1400, 1200, 1250,  1500, 1250,  500,  300,  350,     0,  200, 1800, 1400,  1200, 1250, 1300],
    "SRN":  [1600, 1680, 2400, 2500, 1200,  950, 1550, 1350,   700, 950,    950, 1450, 1650, 1050,   950,  850, 1350, 1350, 1350,  1450,  1450, 1250, 1300,  1550, 1300,  600,  400,  450,  200,    0, 1900, 1500,  1300, 1350, 1450],
    "DEL":  [ 550,  620, 1750, 1850,  500, 1700, 2300, 2100,  1700, 1400,  1500, 2200, 2400, 2200,  2100, 2000, 1800, 1800, 900,   2100,  600,  400,  450,   700,  500,  1700, 1800, 1700, 1800, 1900,    0,  900,   1000,  850, 2200],
    "JAL":  [ 950, 1020, 1750, 1850,  700, 1650, 2150, 1950,  1400, 1100,  1150, 2050, 2250, 1750,  1650, 1550, 1800, 1800, 1050,  1950,  600,  400,  450,   700,  500,  1250, 1400, 1300, 1400, 1500,  900,    0,   650,  550, 2050],
    "SURAT":[1200, 1260, 2000, 2100,  700, 1350, 1900, 1650,  1100, 900,    1000, 1750, 1950, 1300,  1200, 1100, 1500, 1500,  600,  1600,  600,  400,  450,   700,  550,  1050, 1200, 1150, 1200, 1300, 1000,  650,     0,  300, 1750],
    "GAZ":  [1050, 1130, 1850, 1950,  750, 1400, 1900, 1700,  1300, 950,   1000, 1800, 2000, 1400,  1300, 1200, 1600, 1600, 1000,  1700,  550,  350,  400,   650,  500,  1050, 1250, 1100, 1250, 1350,  850,  550,   300,    0, 1800],
    "BHAGIRATH":[1800,1870,2600,2700,1350, 900,1500,1300, 1350, 1600, 1550, 1500, 1700, 1800, 1700, 1600, 2000, 2000, 1600, 1300, 1900, 1700, 1750, 2050, 1800, 1500, 1300, 1100, 1300, 1450, 2200, 2050, 1750, 1800,    0],
}

locations = [
    "MW1", "MW2", "NEL", "LUH", "MYSRD", "BAL", "PHUR", "BUDH", "MUM", "NAG",
    "MILL_RD", "SAN", "CHN", "MUD", "GOL_EXT", "OT", "KOL_CITY", "KOL_EXT",
    "AMD", "HYD", "JPR", "RAP", "RAP_EXT", "RAN", "BPL", "GUT", "CUK", "LUK",
    "PAT", "SRN", "DEL", "JAL", "SURAT", "GAZ", "BHAGIRATH"
]


DIST = pd.DataFrame(dist_data, index=locations, columns=locations)

RATE_PER_KM = 45

# ==========================================================
# ALLOWED DEPOT TRANSFER PAIRS  (distance cap)
# ==========================================================

MAX_DD_DISTANCE = 400   # km — transfers beyond this are not permitted

allowed_dd_pairs = [
    (i, j)
    for i in depots
    for j in depots
    if i != j and DIST.loc[i, j] <= MAX_DD_DISTANCE
]

print("\nALLOWED DEPOT TRANSFER PAIRS (<=400 km)")
print("-" * 50)
for (i, j) in allowed_dd_pairs:
    print(f"  {i} -> {j}  ({DIST.loc[i, j]:,.0f} km)")

MOQ_DD_CARTONS = {
    (i, j, sku): 1
    for (i, j) in allowed_dd_pairs
    for sku in skus
}

# ==========================================================
# DEPOT CONTRACT TARIFF
# ==========================================================

import pandas as pd

# =============================================================
# ZONE MAP
# =============================================================

zone_map = {

    # --- S1 ---
    "NEL": "S1",
    "MYSRD": "S1",
    "BAL": "S1",
    "CHN": "S1",
    "MUD": "S1",
    "HYD": "S1",
    "MILL_RD": "S1",

    # --- W2 ---
    "PHUR": "W2",
    "BUDH": "W2",
    "MUM": "W2",
    "NAG": "W2",

    # --- W1 ---
    "SAN": "W1",
    "AMD": "W1",
    "SURAT": "W1",

    # --- N1 ---
    "GOL_EXT": "N1",
    "JPR": "N1",
    "GUT": "N1",
    "LUK": "N1",
    "DEL": "N1",
    "GAZ": "N1",

    # --- N2 ---
    "LUH": "N2",
    "SRN": "N2",
    "JAL": "N2",

    # --- E ---
    "KOL_CITY": "E",
    "KOL_EXT": "E",
    "OT": "E",
    "RAN": "E",
    "PAT": "E",
    "CUK": "E",

    # --- CENTRAL ---
    "RAP": "Central",
    "RAP_EXT": "Central",
    "BPL": "Central",
}

# =============================================================
# VOLUMETRIC CONVERSION
# =============================================================

CBM_TO_KG = 167

# =============================================================
# TARIFF TABLE
# =============================================================

tariff = pd.DataFrame([

    # FROM S1
    ["S1", "N1", 9.00, 350, 0.10, 0, 100],
    ["S1", "N2", 10.65, 350, 0.10, 0, 100],
    ["S1", "E", 11.15, 350, 0.10, 0, 100],
    ["S1", "NE", 18.00, 350, 0.10, 0, 100],
    ["S1", "W1", 9.00, 350, 0.10, 0, 100],
    ["S1", "W2", 7.50, 350, 0.10, 0, 100],
    ["S1", "S1", 5.50, 350, 0.10, 0, 100],
    ["S1", "S2", 6.59, 350, 0.10, 0, 100],
    ["S1", "Central", 8.57, 350, 0.10, 0, 100],

    # FROM W2
    ["W2", "N1", 9.00, 350, 0.10, 0, 100],
    ["W2", "N2", 10.50, 350, 0.10, 0, 100],
    ["W2", "E", 12.50, 350, 0.10, 0, 100],
    ["W2", "NE", 18.00, 350, 0.10, 0, 100],
    ["W2", "W1", 5.50, 350, 0.10, 0, 100],
    ["W2", "W2", 5.50, 350, 0.10, 0, 100],
    ["W2", "S1", 9.02, 350, 0.10, 0, 100],
    ["W2", "S2", 11.70, 350, 0.10, 0, 100],
    ["W2", "Central", 7.50, 350, 0.10, 0, 100],

    # FROM N1
    ["N1", "N1", 5.50, 350, 0.10, 0, 100],
    ["N1", "N2", 5.73, 350, 0.10, 0, 100],
    ["N1", "E", 11.50, 350, 0.10, 0, 100],
    ["N1", "NE", 17.69, 350, 0.10, 0, 100],
    ["N1", "W1", 9.00, 350, 0.10, 0, 100],
    ["N1", "W2", 9.00, 350, 0.10, 0, 100],
    ["N1", "S1", 12.50, 350, 0.10, 0, 100],
    ["N1", "S2", 16.00, 350, 0.10, 0, 100],
    ["N1", "Central", 8.39, 350, 0.10, 0, 100],

    # FROM W1
    ["W1", "N1", 8.33, 350, 0.10, 0, 100],
    ["W1", "N2", 9.56, 350, 0.10, 0, 100],
    ["W1", "E", 12.50, 350, 0.10, 0, 100],
    ["W1", "NE", 18.00, 350, 0.10, 0, 100],
    ["W1", "W1", 5.50, 350, 0.10, 0, 100],
    ["W1", "W2", 6.16, 350, 0.10, 0, 100],
    ["W1", "S1", 10.50, 350, 0.10, 0, 100],
    ["W1", "S2", 12.50, 350, 0.10, 0, 100],
    ["W1", "Central", 8.13, 350, 0.10, 0, 100],

    # FROM E
    ["E", "N1", 9.50, 350, 0.10, 0, 100],
    ["E", "N2", 11.00, 350, 0.10, 0, 100],
    ["E", "E", 5.94, 350, 0.10, 0, 100],
    ["E", "NE", 9.50, 350, 0.10, 0, 100],
    ["E", "W1", 9.00, 350, 0.10, 0, 100],
    ["E", "W2", 9.00, 350, 0.10, 0, 100],
    ["E", "S1", 10.50, 350, 0.10, 0, 100],
    ["E", "S2", 12.50, 350, 0.10, 0, 100],
    ["E", "Central", 7.50, 350, 0.10, 0, 100],

    # FROM CENTRAL
    ["Central", "N1", 7.50, 350, 0.10, 0, 100],
    ["Central", "N2", 7.73, 350, 0.10, 0, 100],
    ["Central", "E", 11.00, 350, 0.10, 0, 100],
    ["Central", "NE", 18.00, 350, 0.10, 0, 100],
    ["Central", "W1", 5.50, 350, 0.10, 0, 100],
    ["Central", "W2", 7.50, 350, 0.10, 0, 100],
    ["Central", "S1", 9.50, 350, 0.10, 0, 100],
    ["Central", "S2", 12.50, 350, 0.10, 0, 100],
    ["Central", "Central", 5.78, 350, 0.10, 0, 100],

    # ==========================================================
# FROM N2
# ==========================================================

    ["N2", "N1",       5.73, 350, 0.10, 0, 100],
    ["N2", "N2",       5.50, 350, 0.10, 0, 100],
    ["N2", "E",       11.00, 350, 0.10, 0, 100],
    ["N2", "W1",       9.56, 350, 0.10, 0, 100],
    ["N2", "W2",      10.50, 350, 0.10, 0, 100],
    ["N2", "S1",      12.50, 350, 0.10, 0, 100],
    ["N2", "Central",  7.73, 350, 0.10, 0, 100],

], columns=[
    "FROM_ZONE",
    "TO_ZONE",
    "RATE_PER_KG",
    "MIN_CHARGE",
    "FSC",
    "HANDLING",
    "DOCKET"
])

# ==========================================================
# MAX SALES
# ==========================================================

MAX_SALES = {}
for sku in skus:
    for d in depots:
        MAX_SALES[(sku, d)] = max(sales_history[sku][d])

# ==========================================================
# DERIVED PARAMETERS
# ==========================================================

# Available stock (current + supplier inbound + intransit)
A = {
    (sku, d):
        current_stock[sku][d]
        + supplier_inbound[sku][d]
        + intransit[sku][d]
    for sku in skus
    for d in depots
}

# Freeze targets
F = {
    (sku, d): freeze_stock[sku][d]
    for sku in skus
    for d in depots
}

# ==========================================================
# PROTECTED TRANSFERABLE SURPLUS
# ==========================================================

SUR = {}
for sku in skus:
    for d in depots:
        protection_stock = max(F[(sku, d)], MAX_SALES[(sku, d)])
        SUR[(sku, d)]    = max(A[(sku, d)] - protection_stock, 0)

# ==========================================================
# BUSINESS SHORTAGE PENALTY
# ==========================================================

P = {}
for sku in skus:
    rrs          = sku_master[sku]["rrs"]
    base_penalty = rrs_penalty[rrs]
    for d in depots:
        P[(sku, d)] = base_penalty * depot_weight[d]

# ==========================================================
# DEBUG — TRANSFERABLE SURPLUS CHECK
# ==========================================================

print("\nTRANSFERABLE SURPLUS CHECK")
print("-" * 100)
print(
    f"{'SKU':<18}"
    f"{'Depot':<8}"
    f"{'Avail':>10}"
    f"{'Freeze':>10}"
    f"{'MaxSales':>12}"
    f"{'Protected':>12}"
    f"{'Transferable':>15}"
)
for sku in skus:
    for d in depots:
        protection_stock = max(F[(sku, d)], MAX_SALES[(sku, d)])
        print(
            f"{sku:<18}"
            f"{d:<8}"
            f"{A[(sku, d)]:>10.0f}"
            f"{F[(sku, d)]:>10.0f}"
            f"{MAX_SALES[(sku, d)]:>12.0f}"
            f"{protection_stock:>12.0f}"
            f"{SUR[(sku, d)]:>15.0f}"
        )

# ==========================================================
# DEPOT CONTRACT COST  (per SKU — weight affects freight)
# ==========================================================

def depot_contract_cost(frm, to, sku):
    z1, z2  = zone_map[frm], zone_map[to]
    row     = tariff[
        (tariff.FROM_ZONE == z1) & (tariff.TO_ZONE == z2)
    ].iloc[0]
    sm      = sku_master[sku]
    vol_wt  = cbm_per_carton[sku] * CBM_TO_KG
    wt      = max(sm["actual_wt_carton"], vol_wt)
    freight = max(wt * row.RATE_PER_KG, row.MIN_CHARGE) * (1 + row.FSC)
    fixed   = row.HANDLING + row.DOCKET       # per-shipment activation cost
    var     = freight / sm["units_per_carton"] # per-unit variable cost
    return fixed, var

DD_FIXED = {}
DD_VAR   = {}
for sku in skus:
    for (i, j) in allowed_dd_pairs:
        DD_FIXED[(sku, i, j)], DD_VAR[(sku, i, j)] = \
            depot_contract_cost(i, j, sku)

# ==========================================================
# MW CLUSTERS
# ==========================================================

mw_clusters = {
    "MW1": ["BAL", "HYD", "CHN"],
    "MW2": ["DEL", "MUM"]
}

# ==========================================================
# CANDIDATE ROUTES
# ONLY INSIDE EACH MW CLUSTER
# ==========================================================

mw_clusters = {

    "MW1": [
        "NEL", "MYSRD", "BAL",
        "CHN", "MUD", "HYD", "MILL_RD"
    ],

    "MW2": [
        "PHUR", "BUDH", "MUM", "NAG",
        "SAN", "AMD", "SURAT",

        "DEL", "GAZ", "GUT", "JPR",
        "LUK", "GOL_EXT", "LUH",
        "JAL", "SRN", "BPL",

        "KOL_CITY", "KOL_EXT", "OT",
        "RAN", "PAT", "CUK",
        "RAP", "RAP_EXT"
    ]
}

# ==========================================================
# CANDIDATE ROUTES
# ONLY INSIDE EACH MW CLUSTER
# ==========================================================

candidate_routes = [

    # MW1 routes — remove any route with "NEL" as a stop
    ("MW1", ("MYSRD",)),
    ("MW1", ("BAL",)),
    ("MW1", ("CHN",)),
    ("MW1", ("MUD",)),
    ("MW1", ("HYD",)),
    ("MW1", ("MILL_RD",)),

    ("MW1", ("BAL", "HYD")),
    ("MW1", ("HYD", "BAL")),
    ("MW1", ("BAL", "CHN")),
    ("MW1", ("CHN", "BAL")),
    ("MW1", ("HYD", "CHN")),
    ("MW1", ("CHN", "HYD")),
    ("MW1", ("BAL", "CHN", "HYD")),

    # MW2 routes — remove any route with "LUH" as a stop
    ("MW2", ("PHUR",)),
    ("MW2", ("BUDH",)),
    ("MW2", ("MUM",)),
    ("MW2", ("NAG",)),
    ("MW2", ("SAN",)),
    ("MW2", ("AMD",)),
    ("MW2", ("SURAT",)),
    ("MW2", ("DEL",)),
    ("MW2", ("GAZ",)),
    ("MW2", ("GUT",)),
    ("MW2", ("JPR",)),
    ("MW2", ("LUK",)),
    ("MW2", ("GOL_EXT",)),
    ("MW2", ("JAL",)),
    ("MW2", ("SRN",)),
    ("MW2", ("BPL",)),
    ("MW2", ("KOL_CITY",)),
    ("MW2", ("KOL_EXT",)),
    ("MW2", ("OT",)),
    ("MW2", ("RAN",)),
    ("MW2", ("PAT",)),
    ("MW2", ("CUK",)),
    ("MW2", ("RAP",)),
    ("MW2", ("RAP_EXT",)),

    ("MW2", ("PHUR", "BUDH")),
    ("MW2", ("BUDH", "PHUR")),
    ("MW2", ("MUM", "NAG")),
    ("MW2", ("NAG", "MUM")),
    ("MW2", ("SAN", "AMD")),
    ("MW2", ("AMD", "SAN")),
    ("MW2", ("AMD", "SURAT")),
    ("MW2", ("SURAT", "AMD")),
    ("MW2", ("DEL", "GAZ")),
    ("MW2", ("GAZ", "DEL")),
    ("MW2", ("DEL", "JPR")),
    ("MW2", ("JPR", "DEL")),
    ("MW2", ("LUK", "GOL_EXT")),
    ("MW2", ("GOL_EXT", "LUK")),
    ("MW2", ("JAL", "SRN")),
    ("MW2", ("SRN", "JAL")),
    ("MW2", ("KOL_CITY", "KOL_EXT")),
    ("MW2", ("KOL_EXT", "KOL_CITY")),
    ("MW2", ("RAP", "RAP_EXT")),
    ("MW2", ("RAP_EXT", "RAP")),
    ("MW2", ("PAT", "RAN")),
    ("MW2", ("RAN", "PAT")),
    ("MW2", ("OT", "CUK")),
    ("MW2", ("CUK", "OT")),

    ("MW2", ("PHUR", "MUM", "NAG")),
    ("MW2", ("SAN", "AMD", "SURAT")),
    ("MW2", ("DEL", "GAZ", "JPR")),
    ("MW2", ("LUK", "GOL_EXT", "DEL")),
    ("MW2", ("KOL_CITY", "PAT", "RAN")),
    ("MW2", ("RAP", "RAP_EXT", "CUK")),
]

# ==========================================================
# ROUTE DISTANCE
# ==========================================================

def route_distance(warehouse, depot_sequence):
    stops = [warehouse] + list(depot_sequence)
    return sum(
        DIST.loc[stops[k], stops[k + 1]]
        for k in range(len(stops) - 1)
    )

# ==========================================================
# CREATE ROUTES
# ==========================================================

routes = []

for idx, (w, stops) in enumerate(candidate_routes):
    dist = route_distance(w, stops)
    routes.append({
        "id"         : idx,
        "warehouse"  : w,
        "stops"      : stops,
        "n_stops"    : len(stops),
        "dist_km"    : dist,
        "travel_cost": dist * RATE_PER_KM,
    })

R = len(routes)

print("\nCLUSTER BASED ROUTES")
print("-" * 50)
for r in routes:
    print(f"  {r['warehouse']} -> " + " -> ".join(r["stops"]))
print(f"\n  Total candidate routes : {R}")

# ==========================================================
# STOCK SNAPSHOT
# ==========================================================

print("\nSTOCK SNAPSHOT")
print(
    f"{'SKU':<18} {'Depot':<6} {'Avail':>7} {'Freeze':>7} "
    f"{'Surplus':>8} {'Gap':>7} {'Penalty ₹/unit':>16}"
)
for sku in skus:
    for d in depots:
        gap = A[(sku, d)] - F[(sku, d)]
        print(
            f"{sku:<18} {d:<6} {A[(sku, d)]:>7.0f} {F[(sku, d)]:>7.0f} "
            f"{SUR[(sku, d)]:>8.0f} {gap:>+7.0f} {P[(sku, d)]:>16,.0f}"
        )
    print()
```

```python


# ==========================================================
# MODEL
# ==========================================================

print("Building model...")
model = lp.LpProblem("MultiSKU_MultiDrop", lp.LpMinimize)

# ----------------------------------------------------------
# DECISION VARIABLES
# ----------------------------------------------------------

# z[r] = 1 if route r is activated
z = lp.LpVariable.dicts("Route", range(R), cat="Binary")

# q[r, d, sku] = cartons of SKU delivered at stop d on route r
q = lp.LpVariable.dicts(
    "Deliver",
    [
        (r, d, sku)
        for r in range(R)
        for d in routes[r]["stops"]
        for sku in skus
    ],
    lowBound=0,
    cat="Integer"
)

# Truck type per route (exactly one type per active route)
truck_type = lp.LpVariable.dicts(
    "TruckType",
    [(r, t) for r in range(R) for t in vehicles],
    cat="Binary"
)

# y[sku, i, j] = cartons of SKU transferred depot i → depot j
y = lp.LpVariable.dicts(
    "DD_Cartons",
    [(sku, i, j) for sku in skus for (i, j) in allowed_dd_pairs],
    lowBound=0,
    cat="Integer"
)

# v[sku, i, j] = 1 if depot→depot lane (i→j) active for SKU
v = lp.LpVariable.dicts(
    "Use_DD",
    [(sku, i, j) for sku in skus for (i, j) in allowed_dd_pairs],
    cat="Binary"
)

# s[sku, d] = shortage units at depot d for SKU
s = lp.LpVariable.dicts(
    "Shortage",
    [(sku, d) for sku in skus for d in depots],
    lowBound=0
)

# ==========================================================
# OBJECTIVE FUNCTION
# ==========================================================

model += (
    # Truck travel cost (per-km rate × distance, shared across all SKUs)
    lp.lpSum(
        routes[r]["travel_cost"] * z[r]
        for r in range(R)
    )
    # Vehicle hire cost
    + lp.lpSum(
        vehicle_hire_cost[t] * truck_type[(r, t)]
        for r in range(R)
        for t in vehicles
    )
    # Depot→Depot contract cost: fixed activation + variable freight per unit
    + lp.lpSum(
        DD_FIXED[(sku, i, j)] * v[(sku, i, j)]
        + DD_VAR[(sku, i, j)] * y[(sku, i, j)] * sku_master[sku]["units_per_carton"]
        for sku in skus
        for (i, j) in allowed_dd_pairs
    )
    # Value-weighted business shortage penalty
    + lp.lpSum(
        P[(sku, d)] * s[(sku, d)]
        for sku in skus
        for d in depots
    )
), "Total_Cost"

# ==========================================================
# CONSTRAINTS
# ==========================================================

# ----------------------------------------------------------
# C1. Depot balance — per SKU per depot
#     Available + inbound (routes + DD) - outbound (DD) + shortage >= freeze
# ----------------------------------------------------------
for sku in skus:
    upc = sku_master[sku]["units_per_carton"]
    for d in depots:
        inbound_routes = lp.lpSum(
            q[(r, d, sku)] * upc
            for r in range(R) if d in routes[r]["stops"]
        )
        inbound_dd  = lp.lpSum(y[(sku, i, d)] * upc for i in depots if (i, d) in allowed_dd_pairs)
        outbound_dd = lp.lpSum(y[(sku, d, j)] * upc for j in depots if (d, j) in allowed_dd_pairs)
        model += (
            A[(sku, d)] + inbound_routes + inbound_dd - outbound_dd + s[(sku, d)]
            >= F[(sku, d)]
        ), f"Balance_{sku}_{d}"

# ----------------------------------------------------------
# C2. Warehouse stock limit — per SKU per warehouse
# ----------------------------------------------------------
for w in warehouses:
    for sku in skus:
        upc = sku_master[sku]["units_per_carton"]
        model += (
            lp.lpSum(
                q[(r, d, sku)] * upc
                for r in range(R) if routes[r]["warehouse"] == w
                for d in routes[r]["stops"]
            )
            <= warehouse_stock[w][sku]
        ), f"WH_Stock_{w}_{sku}"

# ----------------------------------------------------------
# C3. Relay-aware surplus — depot can only send out what it has
#     (own surplus + any inbound relay transfers)
# ----------------------------------------------------------
for sku in skus:
    upc = sku_master[sku]["units_per_carton"]
    for i in depots:
        inbound_relay = lp.lpSum(
            y[(sku, k, i)] * upc for k in depots if (k, i) in allowed_dd_pairs
        )
        model += (
            lp.lpSum(y[(sku, i, j)] * upc for j in depots if (i, j) in allowed_dd_pairs)
            <= SUR[(sku, i)] + inbound_relay
        ), f"Relay_{sku}_{i}"

# ----------------------------------------------------------
# C4. q = 0 if route not active  (tight Big-M per SKU per warehouse)
# ----------------------------------------------------------
for r in range(R):
    w = routes[r]["warehouse"]
    for d in routes[r]["stops"]:
        for sku in skus:
            M_r = math.ceil(
                warehouse_stock[w][sku] / sku_master[sku]["units_per_carton"]
            )
            model += (
                q[(r, d, sku)] <= M_r * z[r]
            ), f"qActive_{r}_{d}_{sku}"

# ----------------------------------------------------------
# C5. MOQ per SKU per route-depot
# ----------------------------------------------------------
for r in range(R):
    w = routes[r]["warehouse"]
    for d in routes[r]["stops"]:
        for sku in skus:
            moq = MOQ_WD_CARTONS[(w, d, sku)]
            model += (
                q[(r, d, sku)] >= moq * z[r]
            ), f"MOQ_{r}_{d}_{sku}"

# ----------------------------------------------------------
# C6. Exactly one truck type per active route
# ----------------------------------------------------------
for r in range(R):
    model += (
        lp.lpSum(truck_type[(r, t)] for t in vehicles) == z[r]
    ), f"OneTruck_{r}"

# ----------------------------------------------------------
# C7. Shared CBM constraint
#     All SKUs loaded on a route must fit within chosen truck
# ----------------------------------------------------------
for r in range(R):
    total_cbm_on_route = lp.lpSum(
        q[(r, d, sku)] * cbm_per_carton[sku]
        for d in routes[r]["stops"]
        for sku in skus
    )
    model += (
        total_cbm_on_route
        <= lp.lpSum(
            vehicle_capacity_cbm[t] * truck_type[(r, t)]
            for t in vehicles
        )
    ), f"SharedCBM_{r}"

# ----------------------------------------------------------
# C8. Depot→Depot MOQ + tight Big-M — per SKU per lane
# ----------------------------------------------------------
total_mw_sku = {
    sku: sum(warehouse_stock[w][sku] for w in warehouses)
    for sku in skus
}

for sku in skus:
    upc = sku_master[sku]["units_per_carton"]
    for (i, j) in allowed_dd_pairs:
        M_dd = math.ceil(
            (SUR[(sku, i)] + total_mw_sku[sku]) / upc
        )
        model += (
            y[(sku, i, j)] >= MOQ_DD_CARTONS[(i, j, sku)] * v[(sku, i, j)]
        ), f"DD_MOQ_lo_{sku}_{i}_{j}"
        model += (
            y[(sku, i, j)] <= M_dd * v[(sku, i, j)]
        ), f"DD_MOQ_hi_{sku}_{i}_{j}"

# ----------------------------------------------------------
# C9. Each depot served by at most ONE route
#     (same route carries all SKUs for that depot)
# ----------------------------------------------------------
for d in depots:
    model += (
        lp.lpSum(z[r] for r in range(R) if d in routes[r]["stops"]) <= 1
    ), f"OneRoute_{d}"

# ==========================================================
# SOLVE
# ==========================================================

print("Solving...")
solver = lp.PULP_CBC_CMD(msg=0, timeLimit=300, gapRel=0.01)
model.solve(solver)

# ==========================================================
# OUTPUT
# ==========================================================

print("\n" + "=" * 75)
print(f"  STATUS     : {lp.LpStatus[model.status]}")
print(f"  TOTAL COST : ₹{lp.value(model.objective):,.2f}")
print("=" * 75)

# ----------------------------------------------------------
# Active routes
# ----------------------------------------------------------
print("\nACTIVE MULTI-DROP ROUTES")
print("-" * 75)

route_cost_total = 0
active_routes    = [r for r in range(R) if (z[r].varValue or 0) > 0.5]

if not active_routes:
    print("  None")
else:
    for r in active_routes:
        w      = routes[r]["warehouse"]
        stops  = routes[r]["stops"]
        dist   = routes[r]["dist_km"]
        travel = routes[r]["travel_cost"]
        used_t = next(
            (t for t in vehicles if (truck_type[(r, t)].varValue or 0) > 0.5), "—"
        )
        hire  = vehicle_hire_cost.get(used_t, 0)
        total = travel + hire
        route_cost_total += total

        # Total CBM across all SKUs on this route
        total_cbm = sum(
            (q[(r, d, sku)].varValue or 0) * cbm_per_carton[sku]
            for d in stops
            for sku in skus
        )
        cap_cbm = vehicle_capacity_cbm.get(used_t, 0)
        util    = round(total_cbm / cap_cbm * 100, 1) if cap_cbm else 0

        print(f"\n  Route    : {w} → " + " → ".join(stops))
        print(
            f"  Truck    : {used_t}  |  Distance: {dist:,.0f} km  |  "
            f"Travel: ₹{travel:,.0f}  |  Hire: ₹{hire:,.0f}  |  Total: ₹{total:,.0f}"
        )
        print(
            f"  CBM used : {total_cbm:.4f} / {cap_cbm} CBM  "
            f"({util}% utilisation)"
        )
        print(
            f"  Cost split equally across {len(stops)} stop(s): "
            f"₹{total / len(stops):,.0f} per stop"
        )

        # Per-stop, per-SKU breakdown
        for d in stops:
            print(f"\n    Stop: {d}")
            print(f"    {'SKU':<18} {'Cartons':>8} {'Units':>8} {'CBM':>8}")
            stop_cbm = 0
            for sku in skus:
                cartons = round(q[(r, d, sku)].varValue or 0)
                if cartons > 0:
                    units_d   = cartons * sku_master[sku]["units_per_carton"]
                    cbm_d     = cartons * cbm_per_carton[sku]
                    stop_cbm += cbm_d
                    print(
                        f"    {sku:<18} {cartons:>8} {units_d:>8} {cbm_d:>8.4f}"
                    )
            print(
                f"    {'Stop CBM total':<18} {'':>8} {'':>8} {stop_cbm:>8.4f}"
            )

print(f"\n  Route subtotal : ₹{route_cost_total:,.0f}")

# ----------------------------------------------------------
# Depot transfers per SKU
# ----------------------------------------------------------
print("\nDEPOT TRANSFERS (per SKU)")
print("-" * 75)

dd_total = 0
any_dd   = False

for sku in skus:
    sku_lines = []
    for (i, j) in allowed_dd_pairs:
        cartons = round(y[(sku, i, j)].varValue or 0)
        if cartons > 0:
            upc     = sku_master[sku]["units_per_carton"]
            units_t = cartons * upc
            fc      = DD_FIXED[(sku, i, j)]
            vc      = DD_VAR[(sku, i, j)] * units_t
            tc      = fc + vc
            dist_ij = DIST.loc[i, j]
            dd_total += tc
            sku_lines.append(
                f"  {i}→{j:<5}  {dist_ij:>5,.0f} km  "
                f"{cartons} cartons ({units_t} units)  "
                f"Fixed:₹{fc:.0f}  Var:₹{vc:.0f}  Total:₹{tc:.0f}"
            )
            any_dd = True
    if sku_lines:
        print(f"\n  SKU: {sku}")
        for line in sku_lines:
            print(line)

if not any_dd:
    print("  None")

print(f"\n  DD subtotal : ₹{dd_total:,.0f}")

# ----------------------------------------------------------
# Shortages
# ----------------------------------------------------------
print("\nSHORTAGES")
print("-" * 75)

shortage_cost = 0
any_short     = False

for sku in skus:
    for d in depots:
        val = round(s[(sku, d)].varValue or 0, 1)
        if val > 0:
            pen = val * P[(sku, d)]
            shortage_cost += pen
            print(
                f"  {sku:<18} {d}  {val} units  "
                f"Penalty: ₹{pen:,.0f}  (₹{P[(sku,d)]:,.0f}/unit)"
            )
            any_short = True

if not any_short:
    print("  None — all SKUs at all depots meet freeze targets ✓")

# ----------------------------------------------------------
# Post-transfer stock per SKU
# ----------------------------------------------------------
print("\nPOST-TRANSFER STOCK POSITION")
print("-" * 75)
print(
    f"  {'SKU':<18} {'Depot':<6} {'Start':>7} {'In(Rt)':>8} "
    f"{'In(DD)':>8} {'Out(DD)':>8} {'Final':>7} {'Freeze':>7} {'Gap':>6}"
)

for sku in skus:
    upc = sku_master[sku]["units_per_carton"]
    for d in depots:
        in_rt  = sum(
            (q[(r, d, sku)].varValue or 0) * upc
            for r in range(R) if d in routes[r]["stops"]
        )
        in_dd  = sum((y[(sku, i, d)].varValue or 0) * upc for i in depots if (i, d) in allowed_dd_pairs)
        out_dd = sum((y[(sku, d, j)].varValue or 0) * upc for j in depots if (d, j) in allowed_dd_pairs)
        final  = A[(sku, d)] + in_rt + in_dd - out_dd
        gap    = final - F[(sku, d)]
        flag   = "✓" if gap >= 0 else "✗"
        print(
            f"  {sku:<18} {d:<6} {A[(sku, d)]:>7.0f} {in_rt:>8.0f} "
            f"{in_dd:>8.0f} {out_dd:>8.0f} {final:>7.0f} {F[(sku, d)]:>7.0f} "
            f"{gap:>+6.0f} {flag}"
        )
    print()

# ----------------------------------------------------------
# CBM utilisation summary
# ----------------------------------------------------------
if active_routes:
    print("\nCBM UTILISATION SUMMARY")
    print("-" * 75)
    print(
        f"  {'Route':<40} {'Truck':<6} {'Used CBM':>10} "
        f"{'Cap CBM':>9} {'Util%':>7}"
    )
    for r in active_routes:
        used_t    = next(
            (t for t in vehicles if (truck_type[(r, t)].varValue or 0) > 0.5), "—"
        )
        total_cbm = sum(
            (q[(r, d, sku)].varValue or 0) * cbm_per_carton[sku]
            for d in routes[r]["stops"]
            for sku in skus
        )
        cap_cbm = vehicle_capacity_cbm.get(used_t, 0)
        util    = round(total_cbm / cap_cbm * 100, 1) if cap_cbm else 0
        rstr    = routes[r]["warehouse"] + "→" + "→".join(routes[r]["stops"])
        print(
            f"  {rstr:<40} {used_t:<6} {total_cbm:>10.4f} "
            f"{cap_cbm:>9} {util:>6.1f}%"
        )

# ==========================================================
# COST SUMMARY
# ==========================================================

print("\nNETWORK COST SUMMARY")
print("-"*75)

total_cost = (
    route_cost_total
    + dd_total
    + shortage_cost
)

# ----------------------------------------------------------
# Logistics Cost
# ----------------------------------------------------------

logistics_cost = (
    route_cost_total
    + dd_total
)

# ----------------------------------------------------------
# PRINT SUMMARY
# ----------------------------------------------------------

print(f"  Primary Distribution Cost   : ₹{route_cost_total:>12,.0f}")

print(f"  Depot Transfer Cost         : ₹{dd_total:>12,.0f}")

print(f"  Total Logistics Cost        : ₹{logistics_cost:>12,.0f}")

# ----------------------------------------------------------
# Optional: show service impact only if shortage exists
# ----------------------------------------------------------

if shortage_cost > 0:

    print(f"  Unserved Demand Impact      : ₹{shortage_cost:>12,.0f}")

print(f"  {'─'*55}")

print(f"  Network Optimization Cost   : ₹{total_cost:>12,.0f}")
```

```python
# ==========================================================
# SENSITIVITY ANALYSIS FRAMEWORK
# ==========================================================

import copy

def compute_model_metrics(
    max_dd_distance,
    rrs_penalty_override,
    demand_shock_factor,
    freeze_factor
):
    """
    Run parameter sweep and return summary metrics.
    Does NOT re-solve — computes observable metrics from data.
    """

    # ── 1. Rebuild allowed pairs under new distance cap ──────
    pairs = [
        (i, j)
        for i in depots for j in depots
        if i != j and DIST.loc[i, j] <= max_dd_distance
    ]

    # ── 2. Rebuild penalties ─────────────────────────────────
    penalties = {}
    for sku in skus:
        rrs = sku_master[sku]["rrs"]
        base = rrs_penalty_override[rrs]
        for d in depots:
            penalties[(sku, d)] = base * depot_weight[d]

    # ── 3. Rebuild MAX_SALES with demand shock ────────────────
    shocked_sales = {}
    for sku in skus:
        for d in depots:
            shocked_sales[(sku, d)] = max(sales_history[sku][d]) * demand_shock_factor

    # ── 4. Rebuild surplus with freeze factor ─────────────────
    surplus = {}
    shortages = {}
    for sku in skus:
        for d in depots:
            frz = freeze_stock[sku][d] * freeze_factor
            protection = max(frz, shocked_sales[(sku, d)])
            surplus[(sku, d)] = max(A[(sku, d)] - protection, 0)

            # Shortage = how much demand exceeds available
            shortages[(sku, d)] = max(shocked_sales[(sku, d)] - A[(sku, d)], 0)

    # ── 5. Compute metrics ────────────────────────────────────

    # Total transferable surplus (units)
    total_surplus = sum(surplus.values())

    # Shortage by RRS category
    shortage_by_rrs = {"Runner": 0, "Repeater": 0, "Stranger": 0}
    for sku in skus:
        rrs = sku_master[sku]["rrs"]
        for d in depots:
            shortage_by_rrs[rrs] += shortages[(sku, d)]

    # Weighted shortage penalty (proxy for objective function cost)
    total_penalty = sum(
        shortages[(sku, d)] * penalties[(sku, d)]
        for sku in skus for d in depots
    )

    # Number of viable transfer pairs
    n_pairs = len(pairs)

    # Estimated logistics cost: pairs × avg distance × rate × avg shipment size
    avg_dist = (
        sum(DIST.loc[i, j] for (i, j) in pairs) / len(pairs)
        if pairs else 0
    )
    # Rough proxy: cost proportional to pairs × avg_dist
    logistics_proxy = n_pairs * avg_dist * RATE_PER_KM / 1_000_000  # in ₹ Lakhs

    return {
        "n_transfer_pairs":   n_pairs,
        "total_surplus":      round(total_surplus),
        "shortage_runner":    round(shortage_by_rrs["Runner"]),
        "shortage_repeater":  round(shortage_by_rrs["Repeater"]),
        "shortage_stranger":  round(shortage_by_rrs["Stranger"]),
        "total_penalty":      round(total_penalty / 1_000_000, 2),  # ₹ Cr
        "logistics_proxy":    round(logistics_proxy, 2),             # ₹ L
    }


# ==========================================================
# SCENARIO 1 — DISTANCE THRESHOLD SENSITIVITY
# ==========================================================

print("\n" + "=" * 80)
print("SENSITIVITY 1 — DISTANCE THRESHOLD (MAX_DD_DISTANCE)")
print("=" * 80)
print(
    f"{'Distance':>12} {'Pairs':>8} {'Logistics(₹L)':>16} "
    f"{'Runner Short':>14} {'Repeater Short':>16} {'Penalty(₹Cr)':>14}"
)
print("-" * 80)

base_metrics = None
for dist_km in [200, 400, 800]:
    m = compute_model_metrics(
        max_dd_distance      = dist_km,
        rrs_penalty_override = rrs_penalty,
        demand_shock_factor  = 1.0,
        freeze_factor        = 1.0
    )
    if base_metrics is None:
        base_metrics = m
        cost_pct = "Baseline"
    else:
        pct = ((m["logistics_proxy"] - base_metrics["logistics_proxy"])
               / base_metrics["logistics_proxy"] * 100) if base_metrics["logistics_proxy"] > 0 else 0
        cost_pct = f"+{pct:.0f}%" if pct >= 0 else f"{pct:.0f}%"

    print(
        f"{dist_km:>10} km "
        f"{m['n_transfer_pairs']:>8} "
        f"{cost_pct:>16} "
        f"{m['shortage_runner']:>14,} "
        f"{m['shortage_repeater']:>16,} "
        f"{m['total_penalty']:>14.2f}"
    )


# ==========================================================
# SCENARIO 2 — SHORTAGE PENALTY WEIGHT SENSITIVITY
# ==========================================================

print("\n" + "=" * 80)
print("SENSITIVITY 2 — SHORTAGE PENALTY WEIGHTS (RRS)")
print("=" * 80)
print(
    f"{'Scenario':<18} {'Runner Pen':>12} {'Penalty(₹Cr)':>14} "
    f"{'Runner Short':>14} {'Stranger Short':>16}"
)
print("-" * 80)

penalty_scenarios = {
    "Low penalty":    {"Runner":  5000, "Repeater": 2000, "Stranger":  500},
    "Medium penalty": {"Runner": 15000, "Repeater": 7000, "Stranger": 2000},
    "High penalty":   {"Runner": 40000, "Repeater":15000, "Stranger": 5000},
}

for label, pen in penalty_scenarios.items():
    m = compute_model_metrics(
        max_dd_distance      = 400,
        rrs_penalty_override = pen,
        demand_shock_factor  = 1.0,
        freeze_factor        = 1.0
    )
    print(
        f"{label:<18} {pen['Runner']:>12,} "
        f"{m['total_penalty']:>14.2f} "
        f"{m['shortage_runner']:>14,} "
        f"{m['shortage_stranger']:>16,}"
    )


# ==========================================================
# SCENARIO 3 — DEMAND SHOCK SENSITIVITY
# ==========================================================

print("\n" + "=" * 80)
print("SENSITIVITY 3 — DEMAND SHOCK FACTOR")
print("=" * 80)
print(
    f"{'Shock':>10} {'Runner Short':>14} {'Repeater Short':>16} "
    f"{'Stranger Short':>16} {'Penalty(₹Cr)':>14}"
)
print("-" * 80)

for shock in [0.75, 1.0, 1.25, 1.5, 2.0]:
    m = compute_model_metrics(
        max_dd_distance      = 400,
        rrs_penalty_override = rrs_penalty,
        demand_shock_factor  = shock,
        freeze_factor        = 1.0
    )
    print(
        f"{shock:>9.0%} "
        f"{m['shortage_runner']:>14,} "
        f"{m['shortage_repeater']:>16,} "
        f"{m['shortage_stranger']:>16,} "
        f"{m['total_penalty']:>14.2f}"
    )


# ==========================================================
# SCENARIO 4 — FREEZE STOCK FACTOR (Sales Protection Logic)
# ==========================================================

print("\n" + "=" * 80)
print("SENSITIVITY 4 — FREEZE STOCK FACTOR (Sales Protection)")
print("=" * 80)
print(
    f"{'Freeze Factor':>14} {'Label':<22} {'Total Surplus':>14} "
    f"{'Runner Short':>14} {'Penalty(₹Cr)':>14}"
)
print("-" * 80)

freeze_scenarios = {
    0.0:  "No protection",
    0.5:  "Half protection",
    1.0:  "Full protection (base)",
    1.5:  "150% protection",
}

for factor, label in freeze_scenarios.items():
    m = compute_model_metrics(
        max_dd_distance      = 400,
        rrs_penalty_override = rrs_penalty,
        demand_shock_factor  = 1.0,
        freeze_factor        = factor
    )
    print(
        f"{factor:>13.1f}x "
        f"{label:<22} "
        f"{m['total_surplus']:>14,} "
        f"{m['shortage_runner']:>14,} "
        f"{m['total_penalty']:>14.2f}"
    )


# ==========================================================
# SCENARIO 5 — COMBINED WORST CASE vs BEST CASE
# ==========================================================

print("\n" + "=" * 80)
print("SENSITIVITY 5 — BEST CASE vs WORST CASE")
print("=" * 80)

combined_scenarios = {
    "Best case":  dict(max_dd_distance=800, rrs_penalty_override={"Runner":40000,"Repeater":15000,"Stranger":5000}, demand_shock_factor=0.75, freeze_factor=0.5),
    "Base case":  dict(max_dd_distance=400, rrs_penalty_override=rrs_penalty, demand_shock_factor=1.0,  freeze_factor=1.0),
    "Worst case": dict(max_dd_distance=200, rrs_penalty_override={"Runner":5000,"Repeater":2000,"Stranger":500},  demand_shock_factor=1.5,  freeze_factor=1.5),
}

print(f"{'Scenario':<14} {'Pairs':>8} {'Surplus':>10} {'Runner Short':>14} {'Penalty(₹Cr)':>14}")
print("-" * 65)
for label, params in combined_scenarios.items():
    m = compute_model_metrics(**params)
    print(
        f"{label:<14} "
        f"{m['n_transfer_pairs']:>8} "
        f"{m['total_surplus']:>10,} "
        f"{m['shortage_runner']:>14,} "
        f"{m['total_penalty']:>14.2f}"
    )
```
